## Data clustering

Perform K-means clustering on the prepared dataset.


**Reasoning**:
Perform K-means clustering on the prepared dataset by selecting relevant features, instantiating a KMeans object, fitting the model, retrieving the cluster labels, and adding a new column to store the labels.



```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Select relevant features for clustering
features_for_clustering = ['open', 'high', 'low', 'close', 'adj_close', 'volume', 'Rentabilidad', 'daily_return']

# Handle NaN values using SimpleImputer
imputer = SimpleImputer(strategy='mean') # Create an imputer to replace NaNs with the mean
df_prepared[features_for_clustering] = imputer.fit_transform(df_prepared[features_for_clustering]) # Impute NaNs in the selected features

# 2. Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_prepared[features_for_clustering])

# 3. Instantiate a KMeans object
kmeans = KMeans(n_clusters=5, random_state=42)  # Choose an appropriate number of clusters

# 4. Fit the K-means model to the selected features
kmeans.fit(scaled_features)

# 5. Retrieve the cluster labels
cluster_labels = kmeans.labels_

# 6. Add a new column named 'cluster_label' to df_prepared
df_prepared['cluster_label'] = cluster_labels
```

## Feature engineering

### Subtask:
Engineer new features based on the existing ones in `df_prepared`.


**Reasoning**:
Calculate the moving averages, volatility, momentum, and ratio features as instructed.



```python
# Calculate moving averages
df_prepared['ma_5'] = df_prepared.groupby('cluster_label')['adj_close'].transform(lambda x: x.rolling(window=5).mean())
df_prepared['ma_20'] = df_prepared.groupby('cluster_label')['adj_close'].transform(lambda x: x.rolling(window=20).mean())
df_prepared['ma_50'] = df_prepared.groupby('cluster_label')['adj_close'].transform(lambda x: x.rolling(window=50).mean())

# Calculate volatility
df_prepared['volatility_20'] = df_prepared.groupby('cluster_label')['daily_return'].transform(lambda x: x.rolling(window=20).std())

# Calculate momentum
df_prepared['momentum_10'] = df_prepared.groupby('cluster_label')['adj_close'].pct_change(periods=10)

# Calculate ratio features
df_prepared['close_volume_ratio'] = df_prepared['close'] / df_prepared['volume']
df_prepared['high_low_ratio'] = df_prepared['high'] / df_prepared['low']
```

## Model training

### Subtask:
Train a Random Forest Regression model to predict the 'adj_close' price using the features in `df_prepared`.


**Reasoning**:
Train a Random Forest Regression model to predict 'adj_close' using the features in `df_prepared`.



```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Define target variable (y)
y = df_prepared['adj_close']

# Define feature variables (X)
X = df_prepared.drop('adj_close', axis=1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Instantiate a RandomForestRegressor object
rf_model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained Random Forest Regression model in more detail.


**Reasoning**:
Evaluate the trained Random Forest Regression model by generating predictions for the test set, calculating evaluation metrics, and visualizing the results.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Generate predictions for the test set
y_pred = rf_model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")

# Create a scatter plot of actual vs. predicted adj_close values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Adj Close")
plt.ylabel("Predicted Adj Close")
plt.title("Actual vs. Predicted Adj Close")
plt.show()

# Analyze feature importance and visualize it with a bar chart
feature_importances = rf_model.feature_importances_
feature_names = X_train.columns

# Sort feature importances in descending order
sorted_idx = np.argsort(feature_importances)[::-1]
sorted_feature_importances = feature_importances[sorted_idx]
sorted_feature_names = np.array(feature_names)[sorted_idx]

# Create a bar chart to visualize feature importances
plt.figure(figsize=(12, 6))
plt.bar(range(X_train.shape[1]), sorted_feature_importances, align="center")
plt.xticks(range(X_train.shape[1]), sorted_feature_names, rotation=90)
plt.xlabel("Features")
plt.ylabel("Feature Importance")
plt.title("Feature Importance in Random Forest Regression Model")
plt.tight_layout()
plt.show()
```
