```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("/content/actualizados_total.csv")

# Asegurar que la columna de fecha es de tipo datetime
df['price_date'] = pd.to_datetime(df['price_date'])

# Ordenar por fondo y fecha
df = df.sort_values(by=['fund_symbol', 'price_date']).reset_index(drop=True)

# Eliminar la columna "Rentabilidad_Porcentaje"
df = df.drop(columns=['Rentabilidad_Porcentaje'])

# Función para calcular rentabilidades por fondo
def calcular_rentabilidades(grupo):
    grupo = grupo.copy()

    # Rentabilidad diaria
    grupo['rent_diaria'] = grupo['adj_close'].pct_change()

    # Rentabilidad trimestral (~63 días hábiles)
    grupo['rent_trimestre'] = grupo['adj_close'].pct_change(periods=63)

    # Rentabilidad anual (~252 días hábiles)
    grupo['rent_anual'] = grupo['adj_close'].pct_change(periods=252)

    # Rentabilidad acumulada total desde el inicio
    grupo['rent_total_acum'] = (grupo['adj_close'] / grupo['adj_close'].iloc[0]) - 1

    return grupo

# Aplicar la función a cada fondo
df_cohortes = df.groupby('fund_symbol').apply(calcular_rentabilidades).reset_index(drop=True)

# Ordenar el DataFrame por rentabilidad anual de mayor a menor
df_cohortes = df_cohortes.sort_values(by='rent_anual', ascending=False).reset_index(drop=True)

# Crear la columna de "Clasificacion" en base a la rentabilidad anual
df_cohortes['Clasificacion'] = df_cohortes.groupby('price_date')['rent_anual'].rank(ascending=False, method='first')

# Ordenar por fecha y luego por clasificacion
df_cohortes = df_cohortes.sort_values(by=['price_date', 'Clasificacion'], ascending=[True, True]).reset_index(drop=True)

# Guardar el resultado (opcional)
df_cohortes.to_csv("/content/ETF_cohortes.csv", index=False)
```

```python
import pandas as pd

# 1. Asegúrate de que 'price_date' esté en formato datetime
df['price_date'] = pd.to_datetime(df['price_date'])

# 2. Filtrar columnas relevantes
df_filtrado = df[['fund_symbol', 'price_date', 'close']].copy()

# 3. Crear columna de trimestre en formato YYYYQX
df_filtrado['trimestre'] = df_filtrado['price_date'].dt.to_period('Q').astype(str)

# 4. Calcular la mediana por ETF y trimestre
medianas = df_filtrado.groupby(['fund_symbol', 'trimestre'])['close'].median().reset_index()
medianas.rename(columns={'close': 'mediana_close'}, inplace=True)

# 5. Unir la mediana al DataFrame original filtrado
df_con_medianas = df_filtrado.merge(medianas, on=['fund_symbol', 'trimestre'])

# 6. Filtrar filas donde el valor de close coincide con la mediana
df_mediana = df_con_medianas[df_con_medianas['close'] == df_con_medianas['mediana_close']]

# 7. Mostrar resultado (primeros 10 registros)
df_mediana.head(10)
```

```python
# Usar los índices de df_mediana para filtrar el DataFrame original df
df_filtrado_por_mediana = df.loc[df_mediana.index]

# Mostrar primeros registros
df_filtrado_por_mediana.head(100)
```
