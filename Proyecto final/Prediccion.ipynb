```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

1. Carga datos históricos.
2. Selecciona Top N ETFs según NÚMERO DE DATOS. <--- CAMBIO
3. Entrena modelos Prophet.
4. Opcionalmente ejecuta Cross-Validation (controlado por flag).
5. Genera forecasts.
6. Muestra resumen de resultados.
7. Carga datos reales futuros desde un archivo separado. <--- CAMBIO DE ARCHIVO
8. Genera gráficos comparando la predicción con la realidad futura.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector

# Instala Prophet si no lo tienes: pip install prophet
try:
    from prophet import Prophet
    # from prophet.plot import plot_plotly, plot_components_plotly # Descomentar si usas plotly
    from prophet.plot import add_changepoints_to_plot # Importado pero no usado en plot final
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30)
    print("ERROR: La biblioteca 'prophet' no está instalada o completa.")
    print("Por favor, instálala/actualízala ejecutando: pip install prophet")
    print("También puede requerir: pip install cmdstanpy")
    print("*"*30)
    exit()

# Ignorar algunas advertencias comunes
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN ---
TOP_N_ETFS = 5                      # Número de ETFs a procesar
DATASET_VARIABLE_NAME = 'df_top20'  # <<< CONFIRMA NOMBRE VARIABLE DATAFRAME ENTRENAMIENTO >>>
FUTURE_END_DATE = '2025-03-10'      # Fecha final para la proyección
# <<< NOMBRE DEL ARCHIVO CON DATOS REALES ACTUALIZADOS/COMPLETOS >>>
ACTUAL_DATA_FILE = 'ETF_cohortes.csv' # <--- ACTUALIZADO

# --- Control de Cross-Validation ---
EJECUTAR_CROSS_VALIDATION = False # <--- Pon a True si necesitas CV, False para omitirla

# --- Parámetros de Prophet ---
PROPHET_PARAMS = {
    'seasonality_mode': 'multiplicative',
    'changepoint_prior_scale': 1.0,
    'seasonality_prior_scale': 5.0,
    'holidays_prior_scale': 5.0,
    'daily_seasonality': False,
    'weekly_seasonality': True,
    'yearly_seasonality': True,
    'interval_width': 0.80
}
CUSTOM_HOLIDAYS = None

# --- Parámetros de Cross-Validation (Solo se usan si EJECUTAR_CROSS_VALIDATION es True) ---
CV_INITIAL = '730 days'
CV_PERIOD = '90 days'
CV_HORIZON = '180 days'


# --- 2. CARGA Y PREPARACIÓN INICIAL DEL DATAFRAME DE ENTRENAMIENTO ---

# Asegúrate de que df_top20 existe en tu entorno
# Ejemplo si necesitas cargarlo:
# df_top20 = pd.read_csv('tu_archivo_entrenamiento.csv')

if DATASET_VARIABLE_NAME not in globals():
     raise NameError(f"El DataFrame '{DATASET_VARIABLE_NAME}' no se encontró.")
else:
    df_analysis = globals()[DATASET_VARIABLE_NAME].copy()
    print(f"Usando el DataFrame '{DATASET_VARIABLE_NAME}' para entrenamiento.")

# Verificar columnas requeridas para el análisis inicial y entrenamiento
# Ya no necesitamos 'Clasificacion' para la selección, pero sí para el modelo si existiera
required_cols_train = ['fund_symbol', 'price_date', 'adj_close']
if not all(col in df_analysis.columns for col in required_cols_train):
    raise ValueError(f"Faltan columnas requeridas en {DATASET_VARIABLE_NAME}: {required_cols_train}")

# Convertir fecha a datetime y manejar errores
try:
    df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date'])
    print("Columna 'price_date' del dataset de entrenamiento convertida a datetime.")
except Exception as e:
    print(f"Error al convertir 'price_date' en {DATASET_VARIABLE_NAME}: {e}"); exit()


# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---  # <--- SECCIÓN MODIFICADA
print(f"\nIdentificando los {TOP_N_ETFS} ETFs con mayor número de puntos de datos...")

# Calcular el número de puntos (filas) por cada 'fund_symbol'
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')

# Ordenar los ETFs por el conteo de datos en orden descendente (mayor a menor)
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)

# Seleccionar los símbolos de los TOP_N ETFs con más datos
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()

# Validar y mostrar los ETFs seleccionados
if not top_N_etfs_list:
     raise ValueError("No se pudieron contar los datos por ETF o el DataFrame está vacío.")
if len(top_N_etfs_list) < TOP_N_ETFS:
    print(f"Advertencia: Solo se encontraron {len(top_N_etfs_list)} ETFs en el DataFrame.")
    TOP_N_ETFS = len(top_N_etfs_list) # Ajustar N si hay menos de los solicitados

print(f"Se procesarán los siguientes {TOP_N_ETFS} ETFs (seleccionados por tener más datos):")
# Mostrar el conteo con el que fueron seleccionados
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
rank = 1
for index, row in selected_counts_info.iterrows():
     etf_symbol_sel = row['fund_symbol']
     count_sel = row['data_count']
     print(f"  {rank}. {etf_symbol_sel} ({count_sel} puntos)")
     rank += 1
print("-" * 50)
# Nota: Ya no se usa la columna 'Clasificacion' para esta selección.

# --- 4. BUCLE PARA PROCESAR CADA ETF SELECCIONADO ---
results_prophet = {}

for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    results_prophet[etf_symbol] = {'params': PROPHET_PARAMS.copy()}

    # --- 4.1 Filtrar y Preparar Datos de ENTRENAMIENTO ---
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy()
    df_etf = df_etf.sort_values(by='price_date').reset_index(drop=True)
    df_prophet = df_etf[['price_date', 'adj_close']].rename(columns={'price_date': 'ds', 'adj_close': 'y'})
    df_prophet.dropna(subset=['ds', 'y'], inplace=True)

    # --- Calcular si hay suficientes datos para CV ---
    min_data_points_total = 30
    run_cv = True
    try:
        min_data_points_cv_days = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days
        data_range_days = (df_prophet['ds'].max() - df_prophet['ds'].min()).days

        if len(df_prophet) < min_data_points_total:
             print(f"ERROR: Datos de entrenamiento insuficientes ({len(df_prophet)}) para {etf_symbol}. Saltando...")
             results_prophet[etf_symbol]['error'] = "Datos insuficientes para entrenar"
             run_cv = False
             del df_etf, df_prophet; gc.collect()
             continue

        if data_range_days < min_data_points_cv_days:
             print(f"ADVERTENCIA: Rango de datos de entrenamiento ({data_range_days} días) insuficiente para CV. Se omitirá CV.")
             run_cv = False
    except Exception as e_cv_check:
        print(f"Advertencia: No se pudo verificar suficiencia para CV ({e_cv_check}).")
        if len(df_prophet) < min_data_points_total:
             print(f"ERROR: Datos de entrenamiento insuficientes ({len(df_prophet)}). Saltando...")
             results_prophet[etf_symbol]['error'] = "Datos insuficientes para entrenar"
             run_cv = False
             del df_etf, df_prophet; gc.collect()
             continue

    last_real_date = df_prophet['ds'].max()
    results_prophet[etf_symbol]['last_real_date'] = last_real_date
    print(f"Datos de entrenamiento listos. Última fecha: {last_real_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet)}")

    # --- 4.2 Instanciar y Configurar Modelo Prophet ---
    print(f"\nConfigurando modelo Prophet para {etf_symbol}...")
    model = Prophet(holidays=CUSTOM_HOLIDAYS, **PROPHET_PARAMS)
    try:
        model.add_country_holidays(country_name='US')
        print("Feriados de USA añadidos.")
    except Exception as e:
        print(f"Advertencia: No se pudieron añadir feriados de USA: {e}")

    # --- 4.3 Entrenar Modelo ---
    print(f"Entrenando modelo para {etf_symbol}...")
    try:
        model.fit(df_prophet)
        print("Entrenamiento completado.")
    except Exception as e:
        print(f"ERROR al entrenar Prophet para {etf_symbol}: {e}")
        results_prophet[etf_symbol]['error'] = f"Error entrenamiento: {e}"
        del model, df_etf, df_prophet; gc.collect()
        continue

    # --- 4.4 Cross-Validation (Condicional) ---
    df_p = None
    cv_skipped_config_message = "Cross-Validation omitida por configuración (EJECUTAR_CROSS_VALIDATION=False)."
    cv_skipped_data_message = "Cross-Validation omitida (datos de entrenamiento insuficientes)."

    if run_cv and EJECUTAR_CROSS_VALIDATION:
        print(f"\nEjecutando Cross-Validation para {etf_symbol}...")
        print(f"Configuración CV: initial='{CV_INITIAL}', period='{CV_PERIOD}', horizon='{CV_HORIZON}'")
        try:
            df_cv = cross_validation(model, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes")
            df_p = performance_metrics(df_cv, rolling_window=0.1)
            print(f"Cross-Validation completada para {etf_symbol}.")
            results_prophet[etf_symbol]['cv_metrics_summary'] = df_p[['horizon', 'mae', 'rmse', 'mape', 'coverage']].tail().to_dict('records')
            print("--- Métricas de Performance (CV) - Últimos Horizontes ---")
            print(df_p[['horizon', 'mae', 'rmse', 'mape', 'coverage']].tail())
            del df_cv
        except Exception as e:
            print(f"\nERROR durante la Cross-Validation para {etf_symbol}: {e}")
            results_prophet[etf_symbol]['error_cv'] = str(e)
            results_prophet[etf_symbol]['cv_metrics_summary'] = f"Error en CV: {e}"
            df_p = None
    elif not run_cv:
        print(f"\n{cv_skipped_data_message}")
        results_prophet[etf_symbol]['cv_metrics_summary'] = cv_skipped_data_message
        df_p = None
    else:
        print(f"\n{cv_skipped_config_message}")
        results_prophet[etf_symbol]['cv_metrics_summary'] = cv_skipped_config_message
        df_p = None

    # --- 4.5 Generar Forecast Final ---
    print(f"\nGenerando Forecast Final para {etf_symbol} hasta {FUTURE_END_DATE}...")
    forecast = None
    try:
        target_end_date = pd.to_datetime(FUTURE_END_DATE)
        freq = 'B'
        future_dates = pd.date_range(start=last_real_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)
        future_periods = len(future_dates)

        future = model.make_future_dataframe(periods=future_periods, freq=freq)
        future = future[future['ds'] <= target_end_date]

        forecast = model.predict(future)
        forecast['prediction_interval_width'] = forecast['yhat_upper'] - forecast['yhat_lower']
        results_prophet[etf_symbol]['forecast'] = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']].copy()
        print(f"Forecast final generado hasta {forecast['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e:
        print(f"ERROR generando el forecast final para {etf_symbol}: {e}")
        results_prophet[etf_symbol]['error_forecast'] = str(e)
        del model, df_etf, df_prophet, df_p; gc.collect()
        if 'future' in locals(): del future
        if 'forecast' in locals(): del forecast
        continue

    # --- 4.6 Visualización Inicial (Opcional) ---
    print(f"\nGenerando gráficos iniciales para {etf_symbol}...")
    try:
        if forecast is not None:
            fig1 = model.plot(forecast)
            plt.axvline(last_real_date, color='gray', linestyle=':', lw=2, label='Último Dato Entrenamiento')
            plt.title(f'Predicción Prophet Base para {etf_symbol}')
            plt.xlabel("Fecha"); plt.ylabel("Valor (yhat)")
            plt.legend(); plt.tight_layout(); plt.show()

            fig2 = model.plot_components(forecast)
            plt.suptitle(f'Componentes del Modelo Prophet para {etf_symbol}', y=1.02)
            plt.tight_layout(); plt.show()
            print(f"Gráficos iniciales generados.")
        else:
            print("No se generó forecast, omitiendo gráficos iniciales.")
    except Exception as e:
        print(f"Error generando gráficos iniciales: {e}")

    # --- 4.7 Limpieza de Memoria ---
    print(f"Limpiando memoria para {etf_symbol}...")
    del model, df_etf, df_prophet, future, forecast, df_p
    gc.collect()
    print("Limpieza completada.")

    print(f"\n{'='*15} Procesamiento de {etf_symbol} completado {'='*15}")
    print("-" * 70)


print("\n###########################################################")
print("   ANÁLISIS COMPLETADO PARA TODOS LOS ETFs SELECCIONADOS")
print("###########################################################")

# --- 5. MOSTRAR RESUMEN FINAL DE RESULTADOS DEL ANÁLISIS ---
print("\n--- Resumen de Resultados del Análisis ---")
# (Este bloque no necesita cambios)
for etf, data in results_prophet.items():
    print(f"\n--- ETF: {etf} ---")
    print(f"  Parámetros Usados: {data.get('params', 'N/A')}")
    if 'error' in data:
        print(f"  ERROR PRINCIPAL: {data['error']}")
        continue
    print("\n  Resultado Cross-Validation:")
    cv_summary = data.get('cv_metrics_summary')
    if isinstance(cv_summary, str):
        print(f"    {cv_summary}")
    elif isinstance(cv_summary, list) and cv_summary:
        try:
            df_cv_display = pd.DataFrame(cv_summary)
            print(df_cv_display[['horizon', 'mae', 'rmse', 'mape', 'coverage']].round(4).to_string(index=False))
        except Exception as e_disp_cv:
            print(f"    (Error al formatear CV: {e_disp_cv}). Datos:\n    {cv_summary}")
    elif 'error_cv' in data: print(f"    ERROR en CV: {data['error_cv']}")
    else: print("    No se ejecutó o no hay resultados de CV.")
    print("\n  Ejemplo Predicciones Finales (Últimas 5):")
    if 'forecast' in data and not data['forecast'].empty:
        final_preds_display = data['forecast'][['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']].tail()
        print(final_preds_display.round(4).to_string(index=False))
    elif 'error_forecast' in data: print(f"    ERROR en Forecast: {data['error_forecast']}")
    else: print("    No se encontraron datos de forecast.")
    print("-" * 40)
print("\n--- Fin del Resumen del Análisis ---")


# --- 6. COMPARACIÓN CON DATOS REALES ACTUALIZADOS --- # <--- USA NUEVO ARCHIVO

print("\n\n###########################################################")
print(f"      COMPARACIÓN CON DATOS REALES DESDE '{ACTUAL_DATA_FILE}'") # Usa la constante
print("###########################################################")

try:
    # Cargar los datos reales completos/actualizados
    df_actual_new = pd.read_csv(ACTUAL_DATA_FILE) # Usa la constante
    print(f"Datos reales cargados desde '{ACTUAL_DATA_FILE}'.")

    # --- Preparación de Datos Reales ---
    DATE_COL_ACTUAL = 'price_date'
    VALUE_COL_ACTUAL = 'adj_close'
    SYMBOL_COL_ACTUAL = 'fund_symbol'
    required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new.columns for col in required_actual_cols):
        raise ValueError(f"'{ACTUAL_DATA_FILE}' debe contener: {required_actual_cols}")

    df_actual_new[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new[DATE_COL_ACTUAL])
    df_actual_new = df_actual_new.rename(columns={
        DATE_COL_ACTUAL: 'ds',
        VALUE_COL_ACTUAL: 'y_actual',
        SYMBOL_COL_ACTUAL: 'fund_symbol'
    })
    df_actual_new = df_actual_new.sort_values(by=['fund_symbol', 'ds'])
    print("Datos reales preparados (columnas renombradas).")
    comparison_possible = True

except FileNotFoundError:
    print(f"ERROR: El archivo '{ACTUAL_DATA_FILE}' no se encontró.")
    print("Asegúrate de haberlo subido a Google Colab.")
    comparison_possible = False
except Exception as e:
    print(f"Error al cargar o preparar '{ACTUAL_DATA_FILE}': {e}")
    comparison_possible = False

# --- Generar Gráficos de Comparación (MANUALMENTE) ---
if comparison_possible:
    print("\n--- Generando Gráficos de Comparación (Predicción vs Realidad Futura) ---")

    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Comparando para ETF: {etf_symbol} ---")

        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data = results['forecast'].copy()
            last_real_date = results['last_real_date']

            # Filtrar datos reales actualizados para fechas POSTERIORES al entrenamiento
            actual_data_etf = df_actual_new[
                (df_actual_new['fund_symbol'] == etf_symbol) &
                (df_actual_new['ds'] > last_real_date)
            ].copy()

            if actual_data_etf.empty:
                print(f"No se encontraron datos reales en '{ACTUAL_DATA_FILE}' para {etf_symbol} después de {last_real_date.strftime('%Y-%m-%d')}.")
                continue

            print(f"Encontrados {len(actual_data_etf)} puntos reales futuros para comparar.")

            # --- Creación del Gráfico Manual ---
            try:
                print(f"DEBUG: Preparando gráfico manual para {etf_symbol}")
                # Recuperar datos históricos 'y' del df de análisis original
                historical_data_for_plot = df_analysis.loc[
                    df_analysis['fund_symbol'] == etf_symbol, ['price_date', 'adj_close']
                ].rename(columns={'price_date': 'ds', 'adj_close': 'y'})

                comparison_df_base = forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].copy()
                # Merge: forecast + historicos + reales futuros
                comparison_df_with_hist = pd.merge(comparison_df_base, historical_data_for_plot[['ds', 'y']], on='ds', how='left')
                comparison_df_final = pd.merge(comparison_df_with_hist, actual_data_etf[['ds', 'y_actual']], on='ds', how='left')
                print(f"DEBUG: comparison_df_final preparado.")

                # Crear figura y ejes
                fig_comp, ax_comp = plt.subplots(figsize=(12, 6))

                # Graficar elementos
                hist_plot_data = comparison_df_final.dropna(subset=['y'])
                if not hist_plot_data.empty: ax_comp.plot(hist_plot_data['ds'].dt.to_pydatetime(), hist_plot_data['y'], 'k.', label='Observed data points')
                ax_comp.plot(comparison_df_final['ds'].dt.to_pydatetime(), comparison_df_final['yhat'], ls='-', c='#0072B2', label='Forecast')
                ax_comp.fill_between(comparison_df_final['ds'].dt.to_pydatetime(), comparison_df_final['yhat_lower'], comparison_df_final['yhat_upper'], color='#0072B2', alpha=0.2, label='Uncertainty interval')
                actual_plot_data = comparison_df_final.dropna(subset=['y_actual'])
                if not actual_plot_data.empty: ax_comp.plot(actual_plot_data['ds'].dt.to_pydatetime(), actual_plot_data['y_actual'], 'r.', markersize=5, label='Datos Reales (Futuros)')
                else: print(f"Advertencia: No hay datos 'y_actual' válidos para graficar.")
                ax_comp.axvline(last_real_date, color='grey', linestyle=':', lw=2, label='Fin Entrenamiento')

                # Configuración
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
                ax_comp.set_xlabel('Fecha'); ax_comp.set_ylabel('Valor (adj_close)')
                ax_comp.set_title(f'Comparación Predicción vs Realidad para {etf_symbol}')
                days_before = 30
                plot_start_date = last_real_date - pd.Timedelta(days=days_before)
                plot_end_date = comparison_df_final['ds'].max()
                ax_comp.set_xlim([plot_start_date, plot_end_date + pd.Timedelta(days=5)])
                ax_comp.legend()

                # Mostrar
                plt.tight_layout(); plt.show()
                print(f"Gráfico de comparación generado manualmente para {etf_symbol}.")

            except Exception as e_plot:
                print(f"Error generando el gráfico de comparación manual para {etf_symbol}: {e_plot}")
                import traceback
                traceback.print_exc()

        elif 'last_real_date' not in results: print(f"Saltando {etf_symbol}: Falta 'last_real_date'.")
        else: print(f"Saltando {etf_symbol}: No hay 'forecast' disponible.")

        # Limpieza
        if 'forecast_data' in locals(): del forecast_data
        if 'actual_data_etf' in locals(): del actual_data_etf
        if 'comparison_df_final' in locals(): del comparison_df_final
        if 'historical_data_for_plot' in locals(): del historical_data_for_plot
        gc.collect()

    print("\n--- Fin de la Comparación ---")
else:
    print("\nNo se generaron gráficos de comparación debido a problemas al cargar los datos reales.")

print("\n--- Script Finalizado ---")
```

```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

Versión Final v5:
- Selección por Número de Datos.
- Transformación Logarítmica Opcional (con manejo robusto de reversión).
- Regresores Externos Opcionales.
- Ajuste de Hiperparámetros Opcional (vía CV).
- Comparación Final con Datos Reales y Plot Manual (EJE Y LOGARÍTMICO).
- Corrección SyntaxError en limpieza de memoria. <--- CORRECCIÓN
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector
import itertools # Para el grid search
import traceback # Para errores detallados

# Instala Prophet si no lo tienes: pip install prophet cmdstanpy
try:
    from prophet import Prophet
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30); print("ERROR: 'prophet' no instalado. Ejecuta: pip install prophet cmdstanpy"); print("*"*30); exit()

# Ignorar advertencias
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN GENERAL ---
TOP_N_ETFS = 5
DATASET_VARIABLE_NAME = 'df_top20'
FUTURE_END_DATE = '2025-03-10'
ACTUAL_DATA_FILE = 'ETF_cohortes.csv'

# --- 1.1 CONFIGURACIÓN DE MEJORAS ---
USE_LOG_TRANSFORM = True
REGRESSOR_COLUMNS = []
PERFORM_HYPERPARAMETER_TUNING = True # Cambiar a False para saltar tuning
PARAM_GRID = { 'changepoint_prior_scale': [0.01, 0.1, 0.5], 'seasonality_prior_scale': [1.0, 10.0, 20.0], 'holidays_prior_scale': [1.0, 10.0, 20.0] }
PROPHET_BASE_PARAMS = { 'seasonality_mode': 'multiplicative', 'daily_seasonality': False, 'weekly_seasonality': True, 'yearly_seasonality': True, 'interval_width': 0.80 }
CUSTOM_HOLIDAYS = None

# --- 1.2 CONFIGURACIÓN CV ---
EJECUTAR_CROSS_VALIDATION = False # Cambiar a False para saltar CV y Tuning
CV_INITIAL = '730 days'; CV_PERIOD = '90 days'; CV_HORIZON = '180 days'; CV_METRIC_TO_OPTIMIZE = 'rmse'


# --- 2. CARGA Y PREPARACIÓN DATAFRAME ENTRENAMIENTO ---
if DATASET_VARIABLE_NAME not in globals(): raise NameError(f"'{DATASET_VARIABLE_NAME}' no encontrado.")
else: df_analysis = globals()[DATASET_VARIABLE_NAME].copy(); print(f"Usando '{DATASET_VARIABLE_NAME}'.")
required_cols_train = ['fund_symbol', 'price_date', 'adj_close'] + REGRESSOR_COLUMNS
missing_cols = [col for col in required_cols_train if col not in df_analysis.columns]
if missing_cols: raise ValueError(f"Faltan cols en {DATASET_VARIABLE_NAME}: {missing_cols}")
try: df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date']); print("'price_date' entrenamiento a datetime.")
except Exception as e: print(f"Error convirtiendo fecha entreno: {e}"); exit()
df_analysis.dropna(subset=['adj_close'], inplace=True)
if REGRESSOR_COLUMNS:
    for col in REGRESSOR_COLUMNS:
        df_analysis[col] = df_analysis.groupby('fund_symbol')[col].ffill().bfill()
        if df_analysis[col].isnull().any(): print(f"WARN: NaNs regresor entreno '{col}'.")

# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---
print(f"\nIdentificando {TOP_N_ETFS} ETFs con más datos...")
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()
if not top_N_etfs_list: raise ValueError("No se pudo contar datos.");
if len(top_N_etfs_list) < TOP_N_ETFS: print(f"Warn: Solo {len(top_N_etfs_list)} ETFs."); TOP_N_ETFS = len(top_N_etfs_list)
print(f"Se procesarán {TOP_N_ETFS} ETFs (más datos):")
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
for i, (index, row) in enumerate(selected_counts_info.iterrows()): print(f"  {i+1}. {row['fund_symbol']} ({row['data_count']} puntos)")
print("-" * 50)

# --- 4. BUCLE PRINCIPAL PARA PROCESAR CADA ETF ---
results_prophet = {}
for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    results_prophet[etf_symbol] = {'params': {}}; is_log_transformed_etf = False
    # --- 4.1 Prep Datos Entrenamiento ---
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy().sort_values(by='price_date').reset_index(drop=True)
    cols_to_keep = ['price_date', 'adj_close'] + REGRESSOR_COLUMNS
    df_prophet_train = df_etf[cols_to_keep].rename(columns={'price_date': 'ds', 'adj_close': 'y'}).dropna(subset=['ds', 'y'])
    if USE_LOG_TRANSFORM:
        if (df_prophet_train['y'] <= 0).any(): print(f"WARN: Valores <= 0 {etf_symbol}. Saltando Log.")
        else: df_prophet_train['y'] = np.log(df_prophet_train['y']); is_log_transformed_etf = True; print(f"Log Transform aplicada {etf_symbol}.")
    min_data_points_total = 30; can_run_cv = True
    try:
        min_days_cv = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days; data_days = (df_prophet_train['ds'].max() - df_prophet_train['ds'].min()).days
        if len(df_prophet_train) < min_data_points_total: raise ValueError(f"Insuficientes ({len(df_prophet_train)})")
        if data_days < min_days_cv: print(f"WARN: Rango ({data_days} días) corto para CV."); can_run_cv = False
    except Exception as e_check: print(f"ERROR check datos {etf_symbol}: {e_check}"); results_prophet[etf_symbol]['error'] = f"Err check: {e_check}"; del df_etf, df_prophet_train; gc.collect(); continue
    last_train_date = df_prophet_train['ds'].max(); results_prophet[etf_symbol]['last_real_date'] = last_train_date
    print(f"Entren. listo. Última fecha: {last_train_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet_train)}")

    # --- 4.2 Tuning ---
    best_params_for_etf = PROPHET_BASE_PARAMS.copy(); best_run = None
    if PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION and can_run_cv:
        print(f"\n--- Iniciando Tuning {etf_symbol} ---")
        param_combinations = [dict(zip(PARAM_GRID.keys(), v)) for v in itertools.product(*PARAM_GRID.values())]; tuning_results = []
        for params_to_try in param_combinations:
            current_trial_params = {**PROPHET_BASE_PARAMS, **params_to_try}
            try:
                m_tune = Prophet(**current_trial_params, holidays=CUSTOM_HOLIDAYS)
                if REGRESSOR_COLUMNS: [m_tune.add_regressor(r) for r in REGRESSOR_COLUMNS]
                m_tune.fit(df_prophet_train)
                df_cv_tune = cross_validation(m_tune, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes", disable_tqdm=True)
                df_p_tune = performance_metrics(df_cv_tune); metric_value = df_p_tune[CV_METRIC_TO_OPTIMIZE].mean()
                tuning_results.append({'params': params_to_try, 'metric': metric_value})
                del m_tune, df_cv_tune, df_p_tune; gc.collect()
            except Exception as e_tune: print(f"  ERROR CV {params_to_try}: {e_tune}"); tuning_results.append({'params': params_to_try, 'metric': float('inf')})
        if tuning_results:
            best_run = min(tuning_results, key=lambda x: x['metric'])
            if best_run['metric'] != float('inf'): best_params_for_etf.update(best_run['params']); print(f"\nMejores params {etf_symbol}: {best_run['params']} ({CV_METRIC_TO_OPTIMIZE}={best_run['metric']:.4f})")
            else: print("\nTuning falló. Usando base.")
        else: print("\nNo resultados tuning. Usando base.")
    else: # Mensajes de omisión
        if not (PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION): print("\nTuning OMITIDO (desactivado).")
        elif not can_run_cv: print("\nTuning OMITIDO (datos insuficientes).")
    results_prophet[etf_symbol]['params'] = best_params_for_etf.copy(); print(f"Params finales {etf_symbol}: {best_params_for_etf}")

    # --- 4.3 Modelo Final ---
    print(f"\nEntrenando modelo FINAL {etf_symbol}...")
    model_final = Prophet(**best_params_for_etf, holidays=CUSTOM_HOLIDAYS)
    if REGRESSOR_COLUMNS: print(f"Añadiendo regresores: {REGRESSOR_COLUMNS}"); [model_final.add_regressor(r) for r in REGRESSOR_COLUMNS]
    try:
        model_final.fit(df_prophet_train); print("Entrenamiento final completado.")
    except Exception as e_fit_final:
        print(f"ERROR entrenando FINAL: {e_fit_final}")
        results_prophet[etf_symbol]['error'] = f"Err entreno final: {e_fit_final}"
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect();
        if 'model_final' in locals(): del model_final
        continue

    # --- 4.4 Resumen CV/Tuning ---
    if best_run and best_run['metric'] != float('inf') and PERFORM_HYPERPARAMETER_TUNING: results_prophet[etf_symbol]['cv_metrics_summary'] = f"Mejor {CV_METRIC_TO_OPTIMIZE} Tuning: {best_run['metric']:.4f}"
    elif not can_run_cv: results_prophet[etf_symbol]['cv_metrics_summary'] = "CV Omitida (datos insuf.)"
    elif not EJECUTAR_CROSS_VALIDATION: results_prophet[etf_symbol]['cv_metrics_summary'] = "CV Omitida (desactivada)"
    else: results_prophet[etf_symbol]['cv_metrics_summary'] = "CV no ejecutada (sin tuning)"

    # --- 4.5 Forecast Final ---
    print(f"\nGenerando Forecast {etf_symbol}..."); forecast_log_scale = None; future_df_final = None; forecast_final_scale = None
    try:
        target_end_date = pd.to_datetime(FUTURE_END_DATE); freq = 'B'
        future_dates = pd.date_range(start=last_train_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)
        future_df_base = pd.DataFrame({'ds': future_dates}); future_df_final = future_df_base.copy()
        if REGRESSOR_COLUMNS: # ... (lógica regresores futuros) ...
            print("Añadiendo regresores futuros...");
            try: # ... (igual que antes) ...
                df_actual_full=pd.read_csv(ACTUAL_DATA_FILE); df_actual_full['ds']=pd.to_datetime(df_actual_full['price_date'])
                df_reg_fut=df_actual_full.loc[df_actual_full['fund_symbol'] == etf_symbol, ['ds'] + REGRESSOR_COLUMNS].copy()
                future_df_final=pd.merge(future_df_final, df_reg_fut, on='ds', how='left')
                for col in REGRESSOR_COLUMNS:
                    future_df_final[col]=future_df_final[col].ffill().bfill()
                    if future_df_final[col].isnull().any(): last_known = df_prophet_train[col].iloc[-1]; future_df_final[col].fillna(last_known, inplace=True); print(f"WARN: NaNs reg futuro '{col}' rellenados.")
                del df_actual_full, df_reg_fut
            except Exception as e_reg_fut: print(f"ERROR obteniendo regresores futuros: {e_reg_fut}"); raise
        print("Realizando predicción..."); future_df_final = future_df_final[future_df_final['ds'] <= target_end_date];
        if future_df_final.empty: raise ValueError("Future DF vacío.")
        forecast_log_scale = model_final.predict(future_df_final); print("Predicción completada.")
        #print(f"DEBUG: forecast_log_scale tail (yhat PEQUEÑO si log):\n{forecast_log_scale[['ds', 'yhat']].tail()}")
        forecast_final_scale = forecast_log_scale.copy()
        if is_log_transformed_etf:
            #print(f"DEBUG: INTENTANDO REVERTIR Log {etf_symbol}. Flag={is_log_transformed_etf}")
            cols_to_exp=['yhat', 'yhat_lower', 'yhat_upper','trend', 'trend_lower', 'trend_upper']
            cols_rev = []
            for col in cols_to_exp:
                if col in forecast_final_scale.columns:
                    try: forecast_final_scale[col]=np.exp(forecast_final_scale[col]); cols_rev.append(col)
                    except Exception as e_exp: print(f"ERROR np.exp '{col}': {e_exp}")
            #print(f"DEBUG: Columnas revertidas: {cols_rev}")
            #print(f"DEBUG: forecast_final_scale tail (yhat GRANDE ahora):\n{forecast_final_scale[['ds', 'yhat']].tail()}")
            results_prophet[etf_symbol]['log_transform_applied'] = True
        else: #print(f"DEBUG: NO se revierte Log {etf_symbol}.");
             results_prophet[etf_symbol]['log_transform_applied'] = False
        print("Guardando forecast (escala original)...")
        forecast_final_scale['pred_int_width']=forecast_final_scale['yhat_upper'] - forecast_final_scale['yhat_lower']
        cols_save=['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'pred_int_width']; cols_pres=[c for c in cols_save if c in forecast_final_scale.columns]
        results_prophet[etf_symbol]['forecast'] = forecast_final_scale[cols_pres].copy()
        #if 'forecast' in results_prophet[etf_symbol]: print(f"DEBUG: Tail guardado:\n{results_prophet[etf_symbol]['forecast'].tail()}")
        #else: print(f"DEBUG: ERROR - 'forecast' no guardado.")
        print(f"Forecast final guardado. Última fecha: {forecast_final_scale['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e_pred:
        # Bloque except CON CORRECCIÓN de indentación
        print(f"ERROR generando forecast final {etf_symbol}: {e_pred}")
        results_prophet[etf_symbol]['error_forecast'] = str(e_pred)
        traceback.print_exc() # Imprimir traceback completo del error

        # Limpieza robusta - separar los 'del'
        if 'model_final' in locals():
            del model_final
        if 'df_etf' in locals():
            del df_etf
        if 'df_prophet_train' in locals():
            del df_prophet_train
        # <<< Corrección aplicada aquí >>>
        if 'future_df_final' in locals():
            del future_df_final
        if 'forecast_log_scale' in locals():
            del forecast_log_scale
        if 'forecast_final_scale' in locals():
            del forecast_final_scale
        gc.collect()
        # El 'continue' también debe estar al mismo nivel
        continue # Saltar al siguiente ETF

    # --- 4.6 Visualización Inicial (Opcional) ---
    # ...

    # --- 4.7 Limpieza ---
    print(f"Limpiando memoria {etf_symbol}..."); # ... (limpieza igual que antes) ...
    del model_final, df_etf, df_prophet_train
    if 'future_df_final' in locals(): del future_df_final;
    if 'forecast_log_scale' in locals(): del forecast_log_scale;
    if 'forecast_final_scale' in locals(): del forecast_final_scale;
    if 'best_run' in locals(): del best_run;
    if 'tuning_results' in locals(): del tuning_results;
    gc.collect(); print("Limpieza completada.")
    print(f"\n{'='*15} Procesamiento {etf_symbol} completado {'='*15}"); print("-" * 70)

# --- 5. RESUMEN ANÁLISIS ---
print("\n--- Resumen Resultados Análisis ---")
for etf, data in results_prophet.items(): # ... (código resumen sin cambios) ...
    print(f"\n--- ETF: {etf} ---"); print(f"  Params Finales: {data.get('params', 'N/A')}")
    if 'error' in data: print(f"  ERROR: {data['error']}"); continue
    print("\n  Resultado CV/Tuning:"); cv_summary = data.get('cv_metrics_summary'); print(f"    {cv_summary}")
    print("\n  Predicciones Finales (Últimas 5):")
    if 'forecast' in data and not data['forecast'].empty: print(data['forecast'][['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'pred_int_width']].tail().round(4).to_string(index=False))
    elif 'error_forecast' in data: print(f"    ERROR Forecast: {data['error_forecast']}")
    else: print("    No forecast.")
    print("-" * 40)
print("\n--- Fin Resumen Análisis ---")

# --- 6. COMPARACIÓN CON DATOS REALES ---
print("\n\n###########################################################")
print(f"      COMPARACIÓN DATOS REALES DESDE '{ACTUAL_DATA_FILE}'")
print("###########################################################")
try: # ... (carga y preparación datos reales sin cambios) ...
    df_actual_new = pd.read_csv(ACTUAL_DATA_FILE); print(f"Datos reales cargados '{ACTUAL_DATA_FILE}'.")
    DATE_COL_ACTUAL = 'price_date'; VALUE_COL_ACTUAL = 'adj_close'; SYMBOL_COL_ACTUAL = 'fund_symbol'; required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new.columns for col in required_actual_cols): raise ValueError(f"'{ACTUAL_DATA_FILE}' falta cols: {required_actual_cols}")
    df_actual_new[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new[DATE_COL_ACTUAL]); df_actual_new = df_actual_new.rename(columns={DATE_COL_ACTUAL: 'ds', VALUE_COL_ACTUAL: 'y_actual', SYMBOL_COL_ACTUAL: 'fund_symbol'})
    df_actual_new = df_actual_new.sort_values(by=['fund_symbol', 'ds']); df_actual_new.dropna(subset=['ds', 'y_actual'], inplace=True); print("Datos reales preparados."); comparison_possible = True
except FileNotFoundError: print(f"ERROR: '{ACTUAL_DATA_FILE}' no encontrado."); comparison_possible = False
except Exception as e: print(f"Error cargando/prep '{ACTUAL_DATA_FILE}': {e}"); comparison_possible = False

if comparison_possible:
    print("\n--- Generando Gráficos Comparación (MANUAL + EJE Y LOGARÍTMICO) ---") # Título actualizado
    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Comparando para ETF: {etf_symbol} ---")
        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data_plot = results['forecast'].copy(); last_train_date_plot = results['last_real_date']
            actual_data_etf_plot = df_actual_new[(df_actual_new['fund_symbol'] == etf_symbol) & (df_actual_new['ds'] > last_train_date_plot)].copy()
            if actual_data_etf_plot.empty: print(f"No hay datos reales futuros."); continue
            print(f"Encontrados {len(actual_data_etf_plot)} puntos reales futuros.")
            try: # --- Creación Gráfico Manual ---
                historical_data_orig = df_analysis.loc[df_analysis['fund_symbol'] == etf_symbol, ['price_date', 'adj_close']].rename(columns={'price_date': 'ds', 'adj_close': 'y_hist_orig'}).dropna()
                comparison_df_plot = pd.merge(forecast_data_plot[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], actual_data_etf_plot[['ds', 'y_actual']], on='ds', how='left')
                comparison_df_plot = pd.merge(comparison_df_plot, historical_data_orig[['ds', 'y_hist_orig']], on='ds', how='left')
                fig_comp, ax_comp = plt.subplots(figsize=(14, 7))
                hist_plot_data = comparison_df_plot.dropna(subset=['y_hist_orig']); hist_plot_data = hist_plot_data[hist_plot_data['y_hist_orig'] > 0] # Filtrar > 0 para log
                if not hist_plot_data.empty: ax_comp.plot(hist_plot_data['ds'].dt.to_pydatetime(), hist_plot_data['y_hist_orig'], 'k.', label='Observed data points')
                ax_comp.plot(comparison_df_plot['ds'].dt.to_pydatetime(), comparison_df_plot['yhat'], ls='-', c='#0072B2', label='Forecast')
                ax_comp.fill_between(comparison_df_plot['ds'].dt.to_pydatetime(), comparison_df_plot['yhat_lower'], comparison_df_plot['yhat_upper'], color='#0072B2', alpha=0.2, label='Uncertainty interval')
                actual_plot_data = comparison_df_plot.dropna(subset=['y_actual']); actual_plot_data = actual_plot_data[actual_plot_data['y_actual'] > 0] # Filtrar > 0 para log
                if not actual_plot_data.empty: ax_comp.plot(actual_plot_data['ds'].dt.to_pydatetime(), actual_plot_data['y_actual'], 'r.', markersize=5, label='Datos Reales (Futuros)')
                else: print(f"Warn: No 'y_actual' positivos para plot log.")
                ax_comp.axvline(last_train_date_plot, color='grey', linestyle=':', lw=2, label='Fin Entrenamiento')
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2); ax_comp.grid(True, which='minor', c='gray', ls=':', lw=0.5, alpha=0.1)
                ax_comp.set_xlabel('Fecha');
                # <<< MODIFICACIÓN ETIQUETA Y >>>
                ax_comp.set_ylabel('Valor (adj_close) - Escala Logarítmica')
                ax_comp.set_title(f'Comparación Predicción vs Realidad para {etf_symbol} (Escala Log)')
                days_before = 60; plot_start_date = last_train_date_plot - pd.Timedelta(days=days_before); plot_end_date = comparison_df_plot['ds'].max(); ax_comp.set_xlim([plot_start_date, plot_end_date + pd.Timedelta(days=10)])

                # --- ELIMINAR AJUSTE MANUAL DE YLIM ---
                # (Bloque de cálculo de percentiles y set_ylim eliminado)

                # <<< AÑADIR ESTA LÍNEA PARA ESCALA LOG >>>
                ax_comp.set_yscale('log')
                print("DEBUG: Eje Y configurado a escala logarítmica.")

                ax_comp.legend(); plt.tight_layout(); plt.show()
                print(f"Gráfico comparación (log) generado {etf_symbol}.")
            except Exception as e_plot: print(f"Error gráfico {etf_symbol}: {e_plot}"); traceback.print_exc()
        else: # ... (manejo casos sin forecast/fecha) ...
             if 'last_real_date' not in results: print(f"Saltando {etf_symbol}: Falta 'last_real_date'.")
             else: print(f"Saltando {etf_symbol}: No 'forecast'.")
        # Limpieza
        if 'forecast_data_plot' in locals(): del forecast_data_plot;
        if 'actual_data_etf_plot' in locals(): del actual_data_etf_plot;
        if 'comparison_df_plot' in locals(): del comparison_df_plot;
        if 'historical_data_orig' in locals(): del historical_data_orig; gc.collect()
    print("\n--- Fin Comparación ---")
else: print("\nNo se generaron gráficos comparación.")
print("\n--- Script Finalizado ---")
```

# **MODELO PREDICCIONES A FUTURO (HASTA 2030)**

```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

*** VERSIÓN MODIFICADA PARA USAR ETF_cohortes.csv y PREDECIR HASTA 2030 ***

1.  Carga datos directamente desde 'ETF_cohortes.csv'.      ## MODIFICADO ##
2.  Selecciona Top N ETFs según NÚMERO DE DATOS.
3.  Opcionalmente aplica Transformación Logarítmica.
4.  Opcionalmente usa Regresores Externos (con valor futuro constante). ## MODIFICADO ##
5.  Opcionalmente realiza Ajuste de Hiperparámetros (Tuning).
6.  Entrena modelos Prophet (con params base o tuneados).
7.  Genera forecasts hasta 2030-01-01 (revirtiendo log si aplica). ## MODIFICADO ##
8.  Muestra resumen de resultados.
9.  Genera gráficos mostrando histórico y predicción futura.
10. (Sección de comparación mantenida para uso futuro con datos actualizados). ## MODIFICADO ##
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector
import itertools # Para el grid search
import traceback # Para errores detallados

# Instala Prophet si no lo tienes: pip install prophet cmdstanpy
try:
    from prophet import Prophet
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30); print("ERROR: 'prophet' no instalado. Ejecuta: pip install prophet cmdstanpy"); print("*"*30); exit()

# Ignorar advertencias
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN GENERAL ---
TOP_N_ETFS = 5
# DATASET_VARIABLE_NAME = 'df_top20' # ## ELIMINADO ## Ya no se usa variable externa
FUTURE_END_DATE = '2030-01-01'      # ## MODIFICADO ## Fecha final para la proyección
ACTUAL_DATA_FILE = 'ETF_cohortes.csv' # ## MODIFICADO ## Archivo ÚNICO para carga y entrenamiento

# --- 1.1 CONFIGURACIÓN DE MEJORAS ---
USE_LOG_TRANSFORM = True             # Poner a True para usar transformación logarítmica y revertirla
REGRESSOR_COLUMNS = []               # Lista de nombres de columnas a usar como regresores (ej: ['VIX_Close'])
PERFORM_HYPERPARAMETER_TUNING = True # Poner a True para buscar los mejores hiperparámetros vía CV
EJECUTAR_CROSS_VALIDATION = True     # NECESARIO para Tuning. Si es False, se omite Tuning y CV.

# --- 1.2 PARÁMETROS BASE Y DE TUNING ---
PROPHET_BASE_PARAMS = {
    'seasonality_mode': 'multiplicative',
    'daily_seasonality': False,
    'weekly_seasonality': True,
    'yearly_seasonality': True,
    'interval_width': 0.80
}
CUSTOM_HOLIDAYS = None

# --- Parámetros para Grid Search (Solo si PERFORM_HYPERPARAMETER_TUNING es True) ---
PARAM_GRID = {
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [1.0, 10.0, 20.0],
    'holidays_prior_scale': [1.0, 10.0, 20.0] # Asegúrate que CUSTOM_HOLIDAYS esté definido si usas esto
}
CV_METRIC_TO_OPTIMIZE = 'rmse'

# --- Parámetros de Cross-Validation ---
CV_INITIAL = '730 days'
CV_PERIOD = '90 days'
CV_HORIZON = '180 days'


# --- 2. CARGA Y PREPARACIÓN DEL DATAFRAME DE ENTRENAMIENTO --- ## MODIFICADO ##

print(f"Cargando datos directamente desde '{ACTUAL_DATA_FILE}' para entrenamiento...")
try:
    df_analysis = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"Datos cargados exitosamente. Filas: {len(df_analysis)}, Columnas: {len(df_analysis.columns)}")
except FileNotFoundError:
     print(f"ERROR CRÍTICO: El archivo de datos '{ACTUAL_DATA_FILE}' no se encontró.")
     exit()
except Exception as e:
     print(f"ERROR CRÍTICO al cargar '{ACTUAL_DATA_FILE}': {e}")
     exit()

# Verificar columnas requeridas (incluyendo regresores si existen)
required_cols_train = ['fund_symbol', 'price_date', 'adj_close'] + REGRESSOR_COLUMNS
missing_cols = [col for col in required_cols_train if col not in df_analysis.columns]
if missing_cols:
    raise ValueError(f"Faltan columnas requeridas en {ACTUAL_DATA_FILE}: {missing_cols}")

# Convertir fecha a datetime y manejar errores
try:
    df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date'])
    print("Columna 'price_date' convertida a datetime.")
except Exception as e:
    print(f"Error al convertir 'price_date': {e}"); exit()

# Eliminar filas donde el target ('adj_close') es NaN
original_rows = len(df_analysis)
df_analysis.dropna(subset=['adj_close'], inplace=True)
print(f"Filas con 'adj_close' NaN eliminadas. {original_rows - len(df_analysis)} filas eliminadas. Datos restantes: {len(df_analysis)}")

# Pre-procesar NaNs en columnas de regresores (relleno forward/backward por grupo)
if REGRESSOR_COLUMNS:
    print(f"Pre-procesando NaNs en regresores de entrenamiento: {REGRESSOR_COLUMNS}")
    for col in REGRESSOR_COLUMNS:
        original_nan_count = df_analysis[col].isnull().sum()
        if original_nan_count > 0:
            df_analysis[col] = df_analysis.groupby('fund_symbol')[col].transform(lambda x: x.ffill().bfill())
            remaining_nan_count = df_analysis[col].isnull().sum()
            print(f"  - Columna '{col}': {original_nan_count} NaNs iniciales -> {remaining_nan_count} NaNs restantes (rellenados).")
            if remaining_nan_count > 0: print(f"  ADVERTENCIA: Aún quedan {remaining_nan_count} NaNs en '{col}'.")
        df_analysis[col] = pd.to_numeric(df_analysis[col], errors='coerce')
        if df_analysis[col].isnull().any(): print(f"  ADVERTENCIA: Columna '{col}' contiene NaNs después de conversión numérica.")


# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---
# (Sin cambios funcionales, opera sobre el df_analysis cargado)
print(f"\nIdentificando los {TOP_N_ETFS} ETFs con mayor número de puntos de datos en '{ACTUAL_DATA_FILE}'...")
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()
if not top_N_etfs_list: raise ValueError("No se pudieron contar los datos por ETF o el DataFrame está vacío.")
if len(top_N_etfs_list) < TOP_N_ETFS:
    print(f"Advertencia: Solo se encontraron {len(top_N_etfs_list)} ETFs en el DataFrame.")
    TOP_N_ETFS = len(top_N_etfs_list)
print(f"Se procesarán los siguientes {TOP_N_ETFS} ETFs (seleccionados por tener más datos):")
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
for i, (index, row) in enumerate(selected_counts_info.iterrows()):
    print(f"  {i+1}. {row['fund_symbol']} ({row['data_count']} puntos)")
print("-" * 50)


# --- 4. BUCLE PARA PROCESAR CADA ETF SELECCIONADO ---
results_prophet = {}

for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    results_prophet[etf_symbol] = {'params': {}, 'log_transform_applied': False}
    is_log_transformed_etf = False

    # --- 4.1 Filtrar y Preparar Datos de ENTRENAMIENTO ---
    # Filtra df_analysis (que contiene TODO ETF_cohortes.csv) para el ETF actual
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy()
    df_etf = df_etf.sort_values(by='price_date').reset_index(drop=True)

    cols_to_keep = ['price_date', 'adj_close'] + REGRESSOR_COLUMNS
    df_prophet_train = df_etf[cols_to_keep].rename(columns={'price_date': 'ds', 'adj_close': 'y'})
    subset_dropna = ['ds', 'y'] + REGRESSOR_COLUMNS
    original_train_rows = len(df_prophet_train)
    df_prophet_train.dropna(subset=subset_dropna, inplace=True)
    if len(df_prophet_train) < original_train_rows:
         print(f"  Se eliminaron {original_train_rows - len(df_prophet_train)} filas con NaNs en ds/y/regresores para {etf_symbol}.")

    # Aplicar Transformación Logarítmica (Opcional)
    if USE_LOG_TRANSFORM:
        if (df_prophet_train['y'] <= 0).any():
            print(f"ADVERTENCIA: Valores 'y' <= 0 para {etf_symbol}. No se aplicará log.")
            results_prophet[etf_symbol]['log_transform_applied'] = False
        else:
            df_prophet_train['y'] = np.log(df_prophet_train['y'])
            is_log_transformed_etf = True
            results_prophet[etf_symbol]['log_transform_applied'] = True
            print(f"Transformación Logarítmica aplicada a 'y' para {etf_symbol}.")

    # Calcular si hay suficientes datos para CV/Tuning
    min_data_points_total = 30
    can_run_cv_tuning = True
    try:
        if len(df_prophet_train) < min_data_points_total:
            raise ValueError(f"Datos de entrenamiento insuficientes ({len(df_prophet_train)})")
        min_data_points_cv_days = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days
        data_range_days = (df_prophet_train['ds'].max() - df_prophet_train['ds'].min()).days
        if data_range_days < min_data_points_cv_days:
             print(f"ADVERTENCIA: Rango de datos ({data_range_days} días) insuficiente para CV/Tuning. Se omitirá.")
             can_run_cv_tuning = False
    except Exception as e_cv_check:
        print(f"ERROR verificando suficiencia de datos para {etf_symbol}: {e_cv_check}")
        results_prophet[etf_symbol]['error'] = f"Error verificando datos: {e_cv_check}"
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect(); continue # Saltar al siguiente ETF

    last_train_date = df_prophet_train['ds'].max()
    results_prophet[etf_symbol]['last_real_date'] = last_train_date
    print(f"Datos de entrenamiento listos. Última fecha: {last_train_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet_train)}")

    # --- 4.2 Ajuste de Hiperparámetros (Tuning) Opcional ---
    best_params_for_etf = PROPHET_BASE_PARAMS.copy()
    tuning_ran_successfully = False
    best_cv_metric_value = None

    if PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION and can_run_cv_tuning:
        print(f"\n--- Iniciando Tuning para {etf_symbol} ---")
        # (Lógica de tuning sin cambios...)
        all_params = [dict(zip(PARAM_GRID.keys(), v)) for v in itertools.product(*PARAM_GRID.values())]
        tuning_results = []
        print(f"Probando {len(all_params)} combinaciones...")
        for i, params_to_try in enumerate(all_params):
            current_trial_params = {**PROPHET_BASE_PARAMS, **params_to_try}
            # print(f"  Prueba {i+1}/{len(all_params)}: {params_to_try}", end="") # Opcional: más verboso
            try:
                m_tune = Prophet(**current_trial_params, holidays=CUSTOM_HOLIDAYS)
                if REGRESSOR_COLUMNS: [m_tune.add_regressor(r) for r in REGRESSOR_COLUMNS]
                m_tune.fit(df_prophet_train)
                df_cv_tune = cross_validation(m_tune, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes", disable_tqdm=True)
                df_p_tune = performance_metrics(df_cv_tune)
                metric_value = df_p_tune[CV_METRIC_TO_OPTIMIZE].mean()
                tuning_results.append({'params': params_to_try, 'metric': metric_value})
                # print(f" -> {CV_METRIC_TO_OPTIMIZE}={metric_value:.4f}") # Opcional: más verboso
                del m_tune, df_cv_tune, df_p_tune; gc.collect()
            except Exception as e_tune:
                # print(f" -> ERROR: {e_tune}") # Opcional: más verboso
                print(f"  Error en CV prueba {i+1}: {params_to_try} -> {e_tune}") # Mensaje de error menos verboso
                tuning_results.append({'params': params_to_try, 'metric': float('inf')})
                if 'm_tune' in locals(): del m_tune
                if 'df_cv_tune' in locals(): del df_cv_tune
                if 'df_p_tune' in locals(): del df_p_tune
                gc.collect()
        if tuning_results:
            best_run = min(tuning_results, key=lambda x: x['metric'])
            if best_run['metric'] != float('inf'):
                best_params_for_etf.update(best_run['params'])
                best_cv_metric_value = best_run['metric']
                print(f"\nMejores parámetros encontrados: {best_run['params']} ({CV_METRIC_TO_OPTIMIZE}={best_cv_metric_value:.4f})")
                tuning_ran_successfully = True
            else: print("\nTuning falló. Usando parámetros base.")
        else: print("\nNo resultados tuning. Usando parámetros base.")
    else: # Mensajes de omisión
        if not PERFORM_HYPERPARAMETER_TUNING: print("\nTuning OMITIDO (desactivado).")
        elif not EJECUTAR_CROSS_VALIDATION: print("\nTuning OMITIDO (CV global desactivada).")
        elif not can_run_cv_tuning: print("\nTuning OMITIDO (datos insuficientes).")

    results_prophet[etf_symbol]['params'] = best_params_for_etf.copy()
    print(f"Parámetros finales a usar: {best_params_for_etf}")

    # --- 4.3 Entrenar Modelo FINAL ---
    print(f"\nEntrenando modelo FINAL para {etf_symbol}...")
    model_final = Prophet(**best_params_for_etf, holidays=CUSTOM_HOLIDAYS)
    if REGRESSOR_COLUMNS:
        print(f"Añadiendo regresores: {REGRESSOR_COLUMNS}")
        for regressor in REGRESSOR_COLUMNS: model_final.add_regressor(regressor)
    # Añadir feriados USA (opcional)
    # try: ... model_final.add_country_holidays ... except ...

    try:
        model_final.fit(df_prophet_train)
        print("Entrenamiento final completado.")
    except Exception as e_fit_final:
        print(f"ERROR FATAL entrenando FINAL: {e_fit_final}"); traceback.print_exc()
        results_prophet[etf_symbol]['error'] = f"Error entreno final: {e_fit_final}"
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect(); continue

    # --- 4.4 Resumen CV/Tuning ---
    cv_summary_message = "CV/Tuning no ejecutado o fallido."
    if tuning_ran_successfully: cv_summary_message = f"Tuning OK. Mejor {CV_METRIC_TO_OPTIMIZE}: {best_cv_metric_value:.4f}"
    elif not PERFORM_HYPERPARAMETER_TUNING: cv_summary_message = "Tuning desactivado."
    elif not EJECUTAR_CROSS_VALIDATION: cv_summary_message = "Tuning omitido (CV global desactivada)."
    elif not can_run_cv_tuning: cv_summary_message = "Tuning omitido (datos insuficientes)."
    results_prophet[etf_symbol]['cv_metrics_summary'] = cv_summary_message

    # --- 4.5 Generar Forecast Final --- ## LÓGICA REGRESOR MODIFICADA ##
    print(f"\nGenerando Forecast Final para {etf_symbol} hasta {FUTURE_END_DATE}...")
    forecast_final_scale = None; future_df_final = None

    try:
        target_end_date = pd.to_datetime(FUTURE_END_DATE); freq = 'B'
        future_dates = pd.date_range(start=last_train_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)
        if future_dates.empty: raise ValueError(f"No se generaron fechas futuras.")

        future_df_final = pd.DataFrame({'ds': future_dates})
        future_df_final = future_df_final[future_df_final['ds'] <= target_end_date]

        # --- MODIFICADO: Añadir valores futuros para regresores (usando último valor conocido) ---
        if REGRESSOR_COLUMNS:
            print(f"Estableciendo valores futuros para regresores (constantes): {REGRESSOR_COLUMNS}")
            for col in REGRESSOR_COLUMNS:
                if col in df_prophet_train.columns and not df_prophet_train[col].empty:
                    last_known_value = df_prophet_train[col].iloc[-1]
                    future_df_final[col] = last_known_value # Asigna valor constante a toda la columna futura
                    print(f"  - Regressor '{col}' valor futuro fijado a: {last_known_value:.4f}")
                else:
                    future_df_final[col] = 0 # O np.nan o raise error
                    print(f"  - ADVERTENCIA: No se encontró último valor para '{col}'. Futuro fijado a 0.")
            # Verificar NaNs (no debería haber si se asignó constante)
            if future_df_final[REGRESSOR_COLUMNS].isnull().any().any():
                print("  - ADVERTENCIA: NaNs detectados en regresores futuros después de fijar valor. Revise.")

        print(f"Realizando predicción con {len(future_df_final)} fechas futuras...")
        if future_df_final.empty: raise ValueError("DataFrame futuro vacío.")

        forecast_raw = model_final.predict(future_df_final)
        print("Predicción completada.")
        forecast_final_scale = forecast_raw.copy()

        # Revertir Transformación Logarítmica si se aplicó
        if is_log_transformed_etf:
            print(f"Revirtiendo log para {etf_symbol}...")
            cols_to_exp = ['yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']
             # Añadir componentes y regresores si existen
            for col in forecast_raw.columns:
                 if ('seasonality' in col or col in ('weekly','yearly','daily','multiplicative_terms','additive_terms')) and col not in cols_to_exp:
                     cols_to_exp.append(col)
                 if col in REGRESSOR_COLUMNS and col + '_lower' not in cols_to_exp:
                    cols_to_exp.extend([col, col + '_lower', col + '_upper'])

            cols_reverted = []
            for col in cols_to_exp:
                if col in forecast_final_scale.columns:
                    try: forecast_final_scale[col] = np.exp(forecast_final_scale[col]); cols_reverted.append(col)
                    except Exception as e_exp: print(f"  ERROR np.exp '{col}': {e_exp}")
            print(f"Columnas revertidas: {cols_reverted}")

        if 'yhat_upper' in forecast_final_scale.columns and 'yhat_lower' in forecast_final_scale.columns:
             forecast_final_scale['prediction_interval_width'] = forecast_final_scale['yhat_upper'] - forecast_final_scale['yhat_lower']

        cols_to_save = ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']
        cols_present = [c for c in cols_to_save if c in forecast_final_scale.columns]
        results_prophet[etf_symbol]['forecast'] = forecast_final_scale[cols_present].copy()
        print(f"Forecast final guardado. Última fecha: {forecast_final_scale['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e_pred:
        print(f"ERROR FATAL generando forecast: {e_pred}"); traceback.print_exc()
        results_prophet[etf_symbol]['error_forecast'] = str(e_pred)
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        if 'future_df_final' in locals(): del future_df_final
        if 'forecast_raw' in locals(): del forecast_raw
        if 'forecast_final_scale' in locals(): del forecast_final_scale
        gc.collect(); continue

    # --- 4.6 Visualización Inicial (Opcional - Comentada) ---
    # try: ... fig1 = model_final.plot(forecast_raw) ... fig2 = model_final.plot_components(forecast_raw) ...

    # --- 4.7 Limpieza de Memoria ---
    print(f"Limpiando memoria para {etf_symbol}...")
    # (Limpieza sin cambios)
    del model_final, df_etf, df_prophet_train
    if 'future_df_final' in locals(): del future_df_final
    if 'forecast_raw' in locals(): del forecast_raw
    if 'forecast_final_scale' in locals(): del forecast_final_scale
    if 'tuning_results' in locals(): del tuning_results
    if 'best_run' in locals(): del best_run
    if 'all_params' in locals(): del all_params
    gc.collect()
    print("Limpieza completada.")

    print(f"\n{'='*15} Procesamiento de {etf_symbol} completado {'='*15}"); print("-" * 70)


print("\n###########################################################")
print("   ANÁLISIS COMPLETADO PARA TODOS LOS ETFs SELECCIONADOS")
print("###########################################################")

# --- 5. MOSTRAR RESUMEN FINAL DE RESULTADOS DEL ANÁLISIS ---
# (Sin cambios funcionales)
print("\n--- Resumen de Resultados del Análisis ---")
for etf, data in results_prophet.items():
    print(f"\n--- ETF: {etf} ---")
    print(f"  Log Aplicado: {data.get('log_transform_applied', 'N/D')}")
    print(f"  Parámetros Usados: {data.get('params', 'N/A')}")
    if 'error' in data: print(f"  ERROR PRINCIPAL: {data['error']}"); continue
    print("\n  Resultado CV/Tuning:"); print(f"    {data.get('cv_metrics_summary', 'N/D')}")
    print("\n  Predicciones Finales (Últimas 5 - Escala Original):")
    if 'forecast' in data and not data['forecast'].empty:
        cols_display = [c for c in ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width'] if c in data['forecast'].columns]
        if cols_display: print(data['forecast'][cols_display].tail().round(4).to_string(index=False))
        else: print("    No hay columnas estándar de forecast.")
    elif 'error_forecast' in data: print(f"    ERROR Forecast: {data['error_forecast']}")
    else: print("    No forecast.")
    print("-" * 40)
print("\n--- Fin del Resumen del Análisis ---")


# --- 6. COMPARACIÓN CON DATOS REALES (PARA USO FUTURO) --- ## MODIFICADO EN CONCEPTO ##
# NOTA: En esta ejecución, NO habrá puntos rojos ('Datos Reales Posteriores'),
# ya que todo el archivo 'ETF_cohortes.csv' fue usado para entrenar.
# Estos puntos aparecerán si se ejecuta de nuevo el script DESPUÉS de actualizar
# 'ETF_cohortes.csv' con datos posteriores a la última fecha de esta ejecución.

print("\n\n###########################################################")
print(f"      COMPARACIÓN HISTÓRICO vs PREDICCIÓN DESDE '{ACTUAL_DATA_FILE}'") # Título ajustado
print("      (Puntos 'Reales Posteriores' aparecerán en ejecuciones futuras con datos nuevos)")
print("###########################################################")

df_actual_new_for_plot = None # Renombrado para claridad
comparison_possible = False
try:
    # Cargar DE NUEVO los datos para obtener una copia limpia para el plot
    df_actual_new_for_plot = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"\nDatos '{ACTUAL_DATA_FILE}' recargados para gráficos de comparación.")

    DATE_COL_ACTUAL = 'price_date'; VALUE_COL_ACTUAL = 'adj_close'; SYMBOL_COL_ACTUAL = 'fund_symbol'
    required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new_for_plot.columns for col in required_actual_cols):
        raise ValueError(f"'{ACTUAL_DATA_FILE}' recargado debe contener: {required_actual_cols}")

    df_actual_new_for_plot[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new_for_plot[DATE_COL_ACTUAL])
    df_actual_new_for_plot = df_actual_new_for_plot.rename(columns={
        DATE_COL_ACTUAL: 'ds', VALUE_COL_ACTUAL: 'y_actual', SYMBOL_COL_ACTUAL: 'fund_symbol'
    })
    df_actual_new_for_plot.dropna(subset=['ds', 'y_actual', 'fund_symbol'], inplace=True)
    df_actual_new_for_plot = df_actual_new_for_plot.sort_values(by=['fund_symbol', 'ds'])
    print("Datos reales para plot preparados.")
    comparison_possible = True

except FileNotFoundError: print(f"ERROR: '{ACTUAL_DATA_FILE}' no encontrado al recargar."); comparison_possible = False
except Exception as e: print(f"Error cargando/prep '{ACTUAL_DATA_FILE}' para plot: {e}"); traceback.print_exc(); comparison_possible = False

# --- Generar Gráficos de Comparación (Histórico vs Predicción - EJE Y LOG) ---
if comparison_possible and df_actual_new_for_plot is not None:
    print("\n--- Generando Gráficos (Histórico vs Predicción Futura - EJE Y LOG) ---")

    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Graficando para ETF: {etf_symbol} ---")

        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data = results['forecast'].copy()
            last_train_date_plot = results['last_real_date']

            # Filtrar datos reales (del archivo recargado) para ESTE ETF
            # Incluye datos HASTA la última fecha de entrenamiento y, si existieran en el futuro, MÁS ALLÁ
            actual_data_etf_plot = df_actual_new_for_plot[
                (df_actual_new_for_plot['fund_symbol'] == etf_symbol)
            ].copy()

            # Separar datos históricos (usados en entreno) de los 'futuros' (posteriores al entreno)
            historical_data_plot = actual_data_etf_plot[actual_data_etf_plot['ds'] <= last_train_date_plot].rename(columns={'y_actual': 'y_hist_orig'})
            future_actual_data_plot = actual_data_etf_plot[actual_data_etf_plot['ds'] > last_train_date_plot].rename(columns={'y_actual': 'y_future_actual'})

            if historical_data_plot.empty:
                print(f"ADVERTENCIA: No se encontraron datos históricos para {etf_symbol} en el archivo recargado.")
                # Podríamos continuar sólo con el forecast, pero es menos útil
                continue

            if future_actual_data_plot.empty:
                print(f"(Info: No hay datos reales posteriores a {last_train_date_plot.strftime('%Y-%m-%d')} en esta ejecución para comparar)")
            else:
                 print(f"Encontrados {len(future_actual_data_plot)} puntos reales posteriores a la fecha de entrenamiento (para ejecuciones futuras).")


            # --- Creación del Gráfico Manual ---
            try:
                print(f"Preparando gráfico manual para {etf_symbol}")

                # Unir: forecast (predicciones futuras) + datos reales futuros + datos históricos
                # Usar outer merge para asegurar que todas las fechas (históricas y futuras) estén presentes
                comparison_df_plot_full = pd.merge(historical_data_plot[['ds', 'y_hist_orig']],
                                                   forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']],
                                                   on='ds', how='outer')
                comparison_df_plot_full = pd.merge(comparison_df_plot_full,
                                                   future_actual_data_plot[['ds', 'y_future_actual']],
                                                   on='ds', how='outer')

                comparison_df_plot_full = comparison_df_plot_full.sort_values(by='ds').reset_index(drop=True)
                print(f"DataFrame de comparación preparado con {len(comparison_df_plot_full)} filas.")

                fig_comp, ax_comp = plt.subplots(figsize=(14, 7))

                # --- Graficar elementos (con filtro > 0 para log) ---
                # a) Históricos
                hist_plot_data_valid = comparison_df_plot_full.dropna(subset=['y_hist_orig'])
                hist_plot_data_valid = hist_plot_data_valid[hist_plot_data_valid['y_hist_orig'] > 0]
                if not hist_plot_data_valid.empty: ax_comp.plot(hist_plot_data_valid['ds'].dt.to_pydatetime(), hist_plot_data_valid['y_hist_orig'], 'k.', markersize=3, label='Datos Observados (Históricos)')

                # b) Predicción
                forecast_plot_data_valid = comparison_df_plot_full.dropna(subset=['yhat'])
                forecast_plot_data_valid = forecast_plot_data_valid[forecast_plot_data_valid['yhat'] > 0]
                if not forecast_plot_data_valid.empty: ax_comp.plot(forecast_plot_data_valid['ds'].dt.to_pydatetime(), forecast_plot_data_valid['yhat'], ls='-', c='#0072B2', label='Predicción (yhat)')

                # c) Incertidumbre
                uncertainty_plot_data_valid = comparison_df_plot_full.dropna(subset=['yhat_lower', 'yhat_upper'])
                uncertainty_plot_data_valid = uncertainty_plot_data_valid[(uncertainty_plot_data_valid['yhat_lower'] > 0) & (uncertainty_plot_data_valid['yhat_upper'] > 0)]
                if not uncertainty_plot_data_valid.empty: ax_comp.fill_between(uncertainty_plot_data_valid['ds'].dt.to_pydatetime(), uncertainty_plot_data_valid['yhat_lower'], uncertainty_plot_data_valid['yhat_upper'], color='#0072B2', alpha=0.2, label='Intervalo Incertidumbre')

                # d) Reales Futuros (si existen)
                actual_future_plot_data_valid = comparison_df_plot_full.dropna(subset=['y_future_actual'])
                actual_future_plot_data_valid = actual_future_plot_data_valid[actual_future_plot_data_valid['y_future_actual'] > 0]
                if not actual_future_plot_data_valid.empty: ax_comp.plot(actual_future_plot_data_valid['ds'].dt.to_pydatetime(), actual_future_plot_data_valid['y_future_actual'], 'r.', markersize=5, label='Datos Reales (Posteriores Entreno)')
                # No mostrar advertencia si está vacío, ya es esperado en la primera ejecución

                # e) Línea fin entrenamiento
                ax_comp.axvline(last_train_date_plot, color='grey', linestyle=':', lw=2, label=f'Fin Entrenamiento ({last_train_date_plot.strftime("%Y-%m-%d")})')

                # --- Configuración ---
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2); ax_comp.grid(True, which='minor', c='gray', ls=':', lw=0.5, alpha=0.1)
                ax_comp.set_xlabel('Fecha'); ax_comp.set_ylabel('Valor (adj_close) - Escala Logarítmica')
                ax_comp.set_title(f'Histórico y Predicción Futura para {etf_symbol} (Escala Log)')
                ax_comp.set_yscale('log')
                print("Eje Y configurado a escala logarítmica.")

                # Ajustar límites X (desde inicio histórico hasta fin predicción)
                plot_start_date = comparison_df_plot_full['ds'].min()
                plot_end_date = comparison_df_plot_full['ds'].max()
                if pd.notna(plot_start_date) and pd.notna(plot_end_date):
                    ax_comp.set_xlim([plot_start_date - pd.Timedelta(days=30), plot_end_date + pd.Timedelta(days=30)]) # Margen

                ax_comp.legend(); plt.tight_layout(); plt.show()
                print(f"Gráfico generado para {etf_symbol}.")

            except Exception as e_plot:
                print(f"Error generando gráfico para {etf_symbol}: {e_plot}"); traceback.print_exc()

        elif 'last_real_date' not in results: print(f"Saltando gráfico {etf_symbol}: Falta 'last_real_date'.")
        else: print(f"Saltando gráfico {etf_symbol}: No 'forecast'.")

        # Limpieza post-plot
        if 'forecast_data' in locals(): del forecast_data
        if 'actual_data_etf_plot' in locals(): del actual_data_etf_plot
        if 'historical_data_plot' in locals(): del historical_data_plot
        if 'future_actual_data_plot' in locals(): del future_actual_data_plot
        if 'comparison_df_plot_full' in locals(): del comparison_df_plot_full
        if 'hist_plot_data_valid' in locals(): del hist_plot_data_valid
        if 'forecast_plot_data_valid' in locals(): del forecast_plot_data_valid
        if 'uncertainty_plot_data_valid' in locals(): del uncertainty_plot_data_valid
        if 'actual_future_plot_data_valid' in locals(): del actual_future_plot_data_valid
        if 'fig_comp' in locals(): plt.close(fig_comp)
        gc.collect()

    print("\n--- Fin de la Generación de Gráficos ---")
else:
    print("\nNo se generaron gráficos de comparación (problemas al recargar datos o no hay datos válidos).")

# Limpiar el dataframe recargado
if 'df_actual_new_for_plot' in locals(): del df_actual_new_for_plot; gc.collect()

print("\n--- Script Finalizado ---")
```

```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

*** VERSIÓN MODIFICADA PARA USAR ETF_cohortes.csv y PREDECIR HASTA 2030 ***

1.  Carga datos directamente desde 'ETF_cohortes.csv'.      ## MODIFICADO ##
2.  Selecciona Top N ETFs según NÚMERO DE DATOS.
3.  Opcionalmente aplica Transformación Logarítmica.
4.  Opcionalmente usa Regresores Externos (con valor futuro constante). ## MODIFICADO ##
5.  Opcionalmente realiza Ajuste de Hiperparámetros (Tuning).
6.  Entrena modelos Prophet (con params base o tuneados).
7.  Genera forecasts hasta 2030-01-01 (revirtiendo log si aplica). ## MODIFICADO ##
8.  Muestra resumen de resultados.
9.  Genera gráficos mostrando histórico y predicción futura.
10. (Sección de comparación mantenida para uso futuro con datos actualizados). ## MODIFICADO ##
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector
import itertools # Para el grid search
import traceback # Para errores detallados

# Instala Prophet si no lo tienes: pip install prophet cmdstanpy
try:
    from prophet import Prophet
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30); print("ERROR: 'prophet' no instalado. Ejecuta: pip install prophet cmdstanpy"); print("*"*30); exit()

# Ignorar advertencias
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN GENERAL ---
TOP_N_ETFS = 20
# DATASET_VARIABLE_NAME = 'df_top20' # ## ELIMINADO ## Ya no se usa variable externa
FUTURE_END_DATE = '2030-01-01'      # ## MODIFICADO ## Fecha final para la proyección
ACTUAL_DATA_FILE = 'ETF_cohortes.csv' # ## MODIFICADO ## Archivo ÚNICO para carga y entrenamiento

# --- 1.1 CONFIGURACIÓN DE MEJORAS ---
USE_LOG_TRANSFORM = True             # Poner a True para usar transformación logarítmica y revertirla
REGRESSOR_COLUMNS = []               # Lista de nombres de columnas a usar como regresores (ej: ['VIX_Close'])
PERFORM_HYPERPARAMETER_TUNING = True # Poner a True para buscar los mejores hiperparámetros vía CV
EJECUTAR_CROSS_VALIDATION = False     # NECESARIO para Tuning. Si es False, se omite Tuning y CV.

# --- 1.2 PARÁMETROS BASE Y DE TUNING ---
PROPHET_BASE_PARAMS = {
    'seasonality_mode': 'multiplicative',
    'daily_seasonality': False,
    'weekly_seasonality': True,
    'yearly_seasonality': True,
    'interval_width': 0.80
}
CUSTOM_HOLIDAYS = None

# --- Parámetros para Grid Search (Solo si PERFORM_HYPERPARAMETER_TUNING es True) ---
PARAM_GRID = {
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [1.0, 10.0, 20.0],
    'holidays_prior_scale': [1.0, 10.0, 20.0] # Asegúrate que CUSTOM_HOLIDAYS esté definido si usas esto
}
CV_METRIC_TO_OPTIMIZE = 'rmse'

# --- Parámetros de Cross-Validation ---
CV_INITIAL = '730 days'
CV_PERIOD = '90 days'
CV_HORIZON = '180 days'


# --- 2. CARGA Y PREPARACIÓN DEL DATAFRAME DE ENTRENAMIENTO --- ## MODIFICADO ##

print(f"Cargando datos directamente desde '{ACTUAL_DATA_FILE}' para entrenamiento...")
try:
    df_analysis = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"Datos cargados exitosamente. Filas: {len(df_analysis)}, Columnas: {len(df_analysis.columns)}")
except FileNotFoundError:
     print(f"ERROR CRÍTICO: El archivo de datos '{ACTUAL_DATA_FILE}' no se encontró.")
     exit()
except Exception as e:
     print(f"ERROR CRÍTICO al cargar '{ACTUAL_DATA_FILE}': {e}")
     exit()

# Verificar columnas requeridas (incluyendo regresores si existen)
required_cols_train = ['fund_symbol', 'price_date', 'adj_close'] + REGRESSOR_COLUMNS
missing_cols = [col for col in required_cols_train if col not in df_analysis.columns]
if missing_cols:
    raise ValueError(f"Faltan columnas requeridas en {ACTUAL_DATA_FILE}: {missing_cols}")

# Convertir fecha a datetime y manejar errores
try:
    df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date'])
    print("Columna 'price_date' convertida a datetime.")
except Exception as e:
    print(f"Error al convertir 'price_date': {e}"); exit()

# Eliminar filas donde el target ('adj_close') es NaN
original_rows = len(df_analysis)
df_analysis.dropna(subset=['adj_close'], inplace=True)
print(f"Filas con 'adj_close' NaN eliminadas. {original_rows - len(df_analysis)} filas eliminadas. Datos restantes: {len(df_analysis)}")

# Pre-procesar NaNs en columnas de regresores (relleno forward/backward por grupo)
if REGRESSOR_COLUMNS:
    print(f"Pre-procesando NaNs en regresores de entrenamiento: {REGRESSOR_COLUMNS}")
    for col in REGRESSOR_COLUMNS:
        original_nan_count = df_analysis[col].isnull().sum()
        if original_nan_count > 0:
            df_analysis[col] = df_analysis.groupby('fund_symbol')[col].transform(lambda x: x.ffill().bfill())
            remaining_nan_count = df_analysis[col].isnull().sum()
            print(f"  - Columna '{col}': {original_nan_count} NaNs iniciales -> {remaining_nan_count} NaNs restantes (rellenados).")
            if remaining_nan_count > 0: print(f"  ADVERTENCIA: Aún quedan {remaining_nan_count} NaNs en '{col}'.")
        df_analysis[col] = pd.to_numeric(df_analysis[col], errors='coerce')
        if df_analysis[col].isnull().any(): print(f"  ADVERTENCIA: Columna '{col}' contiene NaNs después de conversión numérica.")


# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---
# (Sin cambios funcionales, opera sobre el df_analysis cargado)
print(f"\nIdentificando los {TOP_N_ETFS} ETFs con mayor número de puntos de datos en '{ACTUAL_DATA_FILE}'...")
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()
if not top_N_etfs_list: raise ValueError("No se pudieron contar los datos por ETF o el DataFrame está vacío.")
if len(top_N_etfs_list) < TOP_N_ETFS:
    print(f"Advertencia: Solo se encontraron {len(top_N_etfs_list)} ETFs en el DataFrame.")
    TOP_N_ETFS = len(top_N_etfs_list)
print(f"Se procesarán los siguientes {TOP_N_ETFS} ETFs (seleccionados por tener más datos):")
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
for i, (index, row) in enumerate(selected_counts_info.iterrows()):
    print(f"  {i+1}. {row['fund_symbol']} ({row['data_count']} puntos)")
print("-" * 50)


# --- 4. BUCLE PARA PROCESAR CADA ETF SELECCIONADO ---
results_prophet = {}

for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    results_prophet[etf_symbol] = {'params': {}, 'log_transform_applied': False}
    is_log_transformed_etf = False

    # --- 4.1 Filtrar y Preparar Datos de ENTRENAMIENTO ---
    # Filtra df_analysis (que contiene TODO ETF_cohortes.csv) para el ETF actual
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy()
    df_etf = df_etf.sort_values(by='price_date').reset_index(drop=True)

    cols_to_keep = ['price_date', 'adj_close'] + REGRESSOR_COLUMNS
    df_prophet_train = df_etf[cols_to_keep].rename(columns={'price_date': 'ds', 'adj_close': 'y'})
    subset_dropna = ['ds', 'y'] + REGRESSOR_COLUMNS
    original_train_rows = len(df_prophet_train)
    df_prophet_train.dropna(subset=subset_dropna, inplace=True)
    if len(df_prophet_train) < original_train_rows:
         print(f"  Se eliminaron {original_train_rows - len(df_prophet_train)} filas con NaNs en ds/y/regresores para {etf_symbol}.")

    # Aplicar Transformación Logarítmica (Opcional)
    if USE_LOG_TRANSFORM:
        if (df_prophet_train['y'] <= 0).any():
            print(f"ADVERTENCIA: Valores 'y' <= 0 para {etf_symbol}. No se aplicará log.")
            results_prophet[etf_symbol]['log_transform_applied'] = False
        else:
            df_prophet_train['y'] = np.log(df_prophet_train['y'])
            is_log_transformed_etf = True
            results_prophet[etf_symbol]['log_transform_applied'] = True
            print(f"Transformación Logarítmica aplicada a 'y' para {etf_symbol}.")

    # Calcular si hay suficientes datos para CV/Tuning
    min_data_points_total = 30
    can_run_cv_tuning = True
    try:
        if len(df_prophet_train) < min_data_points_total:
            raise ValueError(f"Datos de entrenamiento insuficientes ({len(df_prophet_train)})")
        min_data_points_cv_days = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days
        data_range_days = (df_prophet_train['ds'].max() - df_prophet_train['ds'].min()).days
        if data_range_days < min_data_points_cv_days:
             print(f"ADVERTENCIA: Rango de datos ({data_range_days} días) insuficiente para CV/Tuning. Se omitirá.")
             can_run_cv_tuning = False
    except Exception as e_cv_check:
        print(f"ERROR verificando suficiencia de datos para {etf_symbol}: {e_cv_check}")
        results_prophet[etf_symbol]['error'] = f"Error verificando datos: {e_cv_check}"
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect(); continue # Saltar al siguiente ETF

    last_train_date = df_prophet_train['ds'].max()
    results_prophet[etf_symbol]['last_real_date'] = last_train_date
    print(f"Datos de entrenamiento listos. Última fecha: {last_train_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet_train)}")

    # --- 4.2 Ajuste de Hiperparámetros (Tuning) Opcional ---
    best_params_for_etf = PROPHET_BASE_PARAMS.copy()
    tuning_ran_successfully = False
    best_cv_metric_value = None

    if PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION and can_run_cv_tuning:
        print(f"\n--- Iniciando Tuning para {etf_symbol} ---")
        # (Lógica de tuning sin cambios...)
        all_params = [dict(zip(PARAM_GRID.keys(), v)) for v in itertools.product(*PARAM_GRID.values())]
        tuning_results = []
        print(f"Probando {len(all_params)} combinaciones...")
        for i, params_to_try in enumerate(all_params):
            current_trial_params = {**PROPHET_BASE_PARAMS, **params_to_try}
            # print(f"  Prueba {i+1}/{len(all_params)}: {params_to_try}", end="") # Opcional: más verboso
            try:
                m_tune = Prophet(**current_trial_params, holidays=CUSTOM_HOLIDAYS)
                if REGRESSOR_COLUMNS: [m_tune.add_regressor(r) for r in REGRESSOR_COLUMNS]
                m_tune.fit(df_prophet_train)
                df_cv_tune = cross_validation(m_tune, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes", disable_tqdm=True)
                df_p_tune = performance_metrics(df_cv_tune)
                metric_value = df_p_tune[CV_METRIC_TO_OPTIMIZE].mean()
                tuning_results.append({'params': params_to_try, 'metric': metric_value})
                # print(f" -> {CV_METRIC_TO_OPTIMIZE}={metric_value:.4f}") # Opcional: más verboso
                del m_tune, df_cv_tune, df_p_tune; gc.collect()
            except Exception as e_tune:
                # print(f" -> ERROR: {e_tune}") # Opcional: más verboso
                print(f"  Error en CV prueba {i+1}: {params_to_try} -> {e_tune}") # Mensaje de error menos verboso
                tuning_results.append({'params': params_to_try, 'metric': float('inf')})
                if 'm_tune' in locals(): del m_tune
                if 'df_cv_tune' in locals(): del df_cv_tune
                if 'df_p_tune' in locals(): del df_p_tune
                gc.collect()
        if tuning_results:
            best_run = min(tuning_results, key=lambda x: x['metric'])
            if best_run['metric'] != float('inf'):
                best_params_for_etf.update(best_run['params'])
                best_cv_metric_value = best_run['metric']
                print(f"\nMejores parámetros encontrados: {best_run['params']} ({CV_METRIC_TO_OPTIMIZE}={best_cv_metric_value:.4f})")
                tuning_ran_successfully = True
            else: print("\nTuning falló. Usando parámetros base.")
        else: print("\nNo resultados tuning. Usando parámetros base.")
    else: # Mensajes de omisión
        if not PERFORM_HYPERPARAMETER_TUNING: print("\nTuning OMITIDO (desactivado).")
        elif not EJECUTAR_CROSS_VALIDATION: print("\nTuning OMITIDO (CV global desactivada).")
        elif not can_run_cv_tuning: print("\nTuning OMITIDO (datos insuficientes).")

    results_prophet[etf_symbol]['params'] = best_params_for_etf.copy()
    print(f"Parámetros finales a usar: {best_params_for_etf}")

    # --- 4.3 Entrenar Modelo FINAL ---
    print(f"\nEntrenando modelo FINAL para {etf_symbol}...")
    model_final = Prophet(**best_params_for_etf, holidays=CUSTOM_HOLIDAYS)
    if REGRESSOR_COLUMNS:
        print(f"Añadiendo regresores: {REGRESSOR_COLUMNS}")
        for regressor in REGRESSOR_COLUMNS: model_final.add_regressor(regressor)
    # Añadir feriados USA (opcional)
    # try: ... model_final.add_country_holidays ... except ...

    try:
        model_final.fit(df_prophet_train)
        print("Entrenamiento final completado.")
    except Exception as e_fit_final:
        print(f"ERROR FATAL entrenando FINAL: {e_fit_final}"); traceback.print_exc()
        results_prophet[etf_symbol]['error'] = f"Error entreno final: {e_fit_final}"
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect(); continue

    # --- 4.4 Resumen CV/Tuning ---
    cv_summary_message = "CV/Tuning no ejecutado o fallido."
    if tuning_ran_successfully: cv_summary_message = f"Tuning OK. Mejor {CV_METRIC_TO_OPTIMIZE}: {best_cv_metric_value:.4f}"
    elif not PERFORM_HYPERPARAMETER_TUNING: cv_summary_message = "Tuning desactivado."
    elif not EJECUTAR_CROSS_VALIDATION: cv_summary_message = "Tuning omitido (CV global desactivada)."
    elif not can_run_cv_tuning: cv_summary_message = "Tuning omitido (datos insuficientes)."
    results_prophet[etf_symbol]['cv_metrics_summary'] = cv_summary_message

    # --- 4.5 Generar Forecast Final --- ## LÓGICA REGRESOR MODIFICADA ##
    print(f"\nGenerando Forecast Final para {etf_symbol} hasta {FUTURE_END_DATE}...")
    forecast_final_scale = None; future_df_final = None

    try:
        target_end_date = pd.to_datetime(FUTURE_END_DATE); freq = 'B'
        future_dates = pd.date_range(start=last_train_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)
        if future_dates.empty: raise ValueError(f"No se generaron fechas futuras.")

        future_df_final = pd.DataFrame({'ds': future_dates})
        future_df_final = future_df_final[future_df_final['ds'] <= target_end_date]

        # --- MODIFICADO: Añadir valores futuros para regresores (usando último valor conocido) ---
        if REGRESSOR_COLUMNS:
            print(f"Estableciendo valores futuros para regresores (constantes): {REGRESSOR_COLUMNS}")
            for col in REGRESSOR_COLUMNS:
                if col in df_prophet_train.columns and not df_prophet_train[col].empty:
                    last_known_value = df_prophet_train[col].iloc[-1]
                    future_df_final[col] = last_known_value # Asigna valor constante a toda la columna futura
                    print(f"  - Regressor '{col}' valor futuro fijado a: {last_known_value:.4f}")
                else:
                    future_df_final[col] = 0 # O np.nan o raise error
                    print(f"  - ADVERTENCIA: No se encontró último valor para '{col}'. Futuro fijado a 0.")
            # Verificar NaNs (no debería haber si se asignó constante)
            if future_df_final[REGRESSOR_COLUMNS].isnull().any().any():
                print("  - ADVERTENCIA: NaNs detectados en regresores futuros después de fijar valor. Revise.")

        print(f"Realizando predicción con {len(future_df_final)} fechas futuras...")
        if future_df_final.empty: raise ValueError("DataFrame futuro vacío.")

        forecast_raw = model_final.predict(future_df_final)
        print("Predicción completada.")
        forecast_final_scale = forecast_raw.copy()

        # Revertir Transformación Logarítmica si se aplicó
        if is_log_transformed_etf:
            print(f"Revirtiendo log para {etf_symbol}...")
            cols_to_exp = ['yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']
             # Añadir componentes y regresores si existen
            for col in forecast_raw.columns:
                 if ('seasonality' in col or col in ('weekly','yearly','daily','multiplicative_terms','additive_terms')) and col not in cols_to_exp:
                     cols_to_exp.append(col)
                 if col in REGRESSOR_COLUMNS and col + '_lower' not in cols_to_exp:
                    cols_to_exp.extend([col, col + '_lower', col + '_upper'])

            cols_reverted = []
            for col in cols_to_exp:
                if col in forecast_final_scale.columns:
                    try: forecast_final_scale[col] = np.exp(forecast_final_scale[col]); cols_reverted.append(col)
                    except Exception as e_exp: print(f"  ERROR np.exp '{col}': {e_exp}")
            print(f"Columnas revertidas: {cols_reverted}")

        if 'yhat_upper' in forecast_final_scale.columns and 'yhat_lower' in forecast_final_scale.columns:
             forecast_final_scale['prediction_interval_width'] = forecast_final_scale['yhat_upper'] - forecast_final_scale['yhat_lower']

        cols_to_save = ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']
        cols_present = [c for c in cols_to_save if c in forecast_final_scale.columns]
        results_prophet[etf_symbol]['forecast'] = forecast_final_scale[cols_present].copy()
        print(f"Forecast final guardado. Última fecha: {forecast_final_scale['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e_pred:
        print(f"ERROR FATAL generando forecast: {e_pred}"); traceback.print_exc()
        results_prophet[etf_symbol]['error_forecast'] = str(e_pred)
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        if 'future_df_final' in locals(): del future_df_final
        if 'forecast_raw' in locals(): del forecast_raw
        if 'forecast_final_scale' in locals(): del forecast_final_scale
        gc.collect(); continue

    # --- 4.6 Visualización Inicial (Opcional - Comentada) ---
    # try: ... fig1 = model_final.plot(forecast_raw) ... fig2 = model_final.plot_components(forecast_raw) ...

    # --- 4.7 Limpieza de Memoria ---
    print(f"Limpiando memoria para {etf_symbol}...")
    # (Limpieza sin cambios)
    del model_final, df_etf, df_prophet_train
    if 'future_df_final' in locals(): del future_df_final
    if 'forecast_raw' in locals(): del forecast_raw
    if 'forecast_final_scale' in locals(): del forecast_final_scale
    if 'tuning_results' in locals(): del tuning_results
    if 'best_run' in locals(): del best_run
    if 'all_params' in locals(): del all_params
    gc.collect()
    print("Limpieza completada.")

    print(f"\n{'='*15} Procesamiento de {etf_symbol} completado {'='*15}"); print("-" * 70)


print("\n###########################################################")
print("   ANÁLISIS COMPLETADO PARA TODOS LOS ETFs SELECCIONADOS")
print("###########################################################")

# --- 5. MOSTRAR RESUMEN FINAL DE RESULTADOS DEL ANÁLISIS ---
# (Sin cambios funcionales)
print("\n--- Resumen de Resultados del Análisis ---")
for etf, data in results_prophet.items():
    print(f"\n--- ETF: {etf} ---")
    print(f"  Log Aplicado: {data.get('log_transform_applied', 'N/D')}")
    print(f"  Parámetros Usados: {data.get('params', 'N/A')}")
    if 'error' in data: print(f"  ERROR PRINCIPAL: {data['error']}"); continue
    print("\n  Resultado CV/Tuning:"); print(f"    {data.get('cv_metrics_summary', 'N/D')}")
    print("\n  Predicciones Finales (Últimas 5 - Escala Original):")
    if 'forecast' in data and not data['forecast'].empty:
        cols_display = [c for c in ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width'] if c in data['forecast'].columns]
        if cols_display: print(data['forecast'][cols_display].tail().round(4).to_string(index=False))
        else: print("    No hay columnas estándar de forecast.")
    elif 'error_forecast' in data: print(f"    ERROR Forecast: {data['error_forecast']}")
    else: print("    No forecast.")
    print("-" * 40)
print("\n--- Fin del Resumen del Análisis ---")


# --- 6. COMPARACIÓN CON DATOS REALES (PARA USO FUTURO) --- ## MODIFICADO EN CONCEPTO ##
# NOTA: En esta ejecución, NO habrá puntos rojos ('Datos Reales Posteriores'),
# ya que todo el archivo 'ETF_cohortes.csv' fue usado para entrenar.
# Estos puntos aparecerán si se ejecuta de nuevo el script DESPUÉS de actualizar
# 'ETF_cohortes.csv' con datos posteriores a la última fecha de esta ejecución.

print("\n\n###########################################################")
print(f"      COMPARACIÓN HISTÓRICO vs PREDICCIÓN DESDE '{ACTUAL_DATA_FILE}'") # Título ajustado
print("      (Puntos 'Reales Posteriores' aparecerán en ejecuciones futuras con datos nuevos)")
print("###########################################################")

df_actual_new_for_plot = None # Renombrado para claridad
comparison_possible = False
try:
    # Cargar DE NUEVO los datos para obtener una copia limpia para el plot
    df_actual_new_for_plot = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"\nDatos '{ACTUAL_DATA_FILE}' recargados para gráficos de comparación.")

    DATE_COL_ACTUAL = 'price_date'; VALUE_COL_ACTUAL = 'adj_close'; SYMBOL_COL_ACTUAL = 'fund_symbol'
    required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new_for_plot.columns for col in required_actual_cols):
        raise ValueError(f"'{ACTUAL_DATA_FILE}' recargado debe contener: {required_actual_cols}")

    df_actual_new_for_plot[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new_for_plot[DATE_COL_ACTUAL])
    df_actual_new_for_plot = df_actual_new_for_plot.rename(columns={
        DATE_COL_ACTUAL: 'ds', VALUE_COL_ACTUAL: 'y_actual', SYMBOL_COL_ACTUAL: 'fund_symbol'
    })
    df_actual_new_for_plot.dropna(subset=['ds', 'y_actual', 'fund_symbol'], inplace=True)
    df_actual_new_for_plot = df_actual_new_for_plot.sort_values(by=['fund_symbol', 'ds'])
    print("Datos reales para plot preparados.")
    comparison_possible = True

except FileNotFoundError: print(f"ERROR: '{ACTUAL_DATA_FILE}' no encontrado al recargar."); comparison_possible = False
except Exception as e: print(f"Error cargando/prep '{ACTUAL_DATA_FILE}' para plot: {e}"); traceback.print_exc(); comparison_possible = False

# --- Generar Gráficos de Comparación (Histórico vs Predicción - EJE Y LOG) ---
if comparison_possible and df_actual_new_for_plot is not None:
    print("\n--- Generando Gráficos (Histórico vs Predicción Futura - EJE Y LOG) ---")

    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Graficando para ETF: {etf_symbol} ---")

        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data = results['forecast'].copy()
            last_train_date_plot = results['last_real_date']

            # Filtrar datos reales (del archivo recargado) para ESTE ETF
            # Incluye datos HASTA la última fecha de entrenamiento y, si existieran en el futuro, MÁS ALLÁ
            actual_data_etf_plot = df_actual_new_for_plot[
                (df_actual_new_for_plot['fund_symbol'] == etf_symbol)
            ].copy()

            # Separar datos históricos (usados en entreno) de los 'futuros' (posteriores al entreno)
            historical_data_plot = actual_data_etf_plot[actual_data_etf_plot['ds'] <= last_train_date_plot].rename(columns={'y_actual': 'y_hist_orig'})
            future_actual_data_plot = actual_data_etf_plot[actual_data_etf_plot['ds'] > last_train_date_plot].rename(columns={'y_actual': 'y_future_actual'})

            if historical_data_plot.empty:
                print(f"ADVERTENCIA: No se encontraron datos históricos para {etf_symbol} en el archivo recargado.")
                # Podríamos continuar sólo con el forecast, pero es menos útil
                continue

            if future_actual_data_plot.empty:
                print(f"(Info: No hay datos reales posteriores a {last_train_date_plot.strftime('%Y-%m-%d')} en esta ejecución para comparar)")
            else:
                 print(f"Encontrados {len(future_actual_data_plot)} puntos reales posteriores a la fecha de entrenamiento (para ejecuciones futuras).")


            # --- Creación del Gráfico Manual ---
            try:
                print(f"Preparando gráfico manual para {etf_symbol}")

                # Unir: forecast (predicciones futuras) + datos reales futuros + datos históricos
                # Usar outer merge para asegurar que todas las fechas (históricas y futuras) estén presentes
                comparison_df_plot_full = pd.merge(historical_data_plot[['ds', 'y_hist_orig']],
                                                   forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']],
                                                   on='ds', how='outer')
                comparison_df_plot_full = pd.merge(comparison_df_plot_full,
                                                   future_actual_data_plot[['ds', 'y_future_actual']],
                                                   on='ds', how='outer')

                comparison_df_plot_full = comparison_df_plot_full.sort_values(by='ds').reset_index(drop=True)
                print(f"DataFrame de comparación preparado con {len(comparison_df_plot_full)} filas.")

                fig_comp, ax_comp = plt.subplots(figsize=(14, 7))

                # --- Graficar elementos (con filtro > 0 para log) ---
                # a) Históricos
                hist_plot_data_valid = comparison_df_plot_full.dropna(subset=['y_hist_orig'])
                hist_plot_data_valid = hist_plot_data_valid[hist_plot_data_valid['y_hist_orig'] > 0]
                if not hist_plot_data_valid.empty: ax_comp.plot(hist_plot_data_valid['ds'].dt.to_pydatetime(), hist_plot_data_valid['y_hist_orig'], 'k.', markersize=3, label='Datos Observados (Históricos)')

                # b) Predicción
                forecast_plot_data_valid = comparison_df_plot_full.dropna(subset=['yhat'])
                forecast_plot_data_valid = forecast_plot_data_valid[forecast_plot_data_valid['yhat'] > 0]
                if not forecast_plot_data_valid.empty: ax_comp.plot(forecast_plot_data_valid['ds'].dt.to_pydatetime(), forecast_plot_data_valid['yhat'], ls='-', c='#0072B2', label='Predicción (yhat)')

                # c) Incertidumbre
                uncertainty_plot_data_valid = comparison_df_plot_full.dropna(subset=['yhat_lower', 'yhat_upper'])
                uncertainty_plot_data_valid = uncertainty_plot_data_valid[(uncertainty_plot_data_valid['yhat_lower'] > 0) & (uncertainty_plot_data_valid['yhat_upper'] > 0)]
                if not uncertainty_plot_data_valid.empty: ax_comp.fill_between(uncertainty_plot_data_valid['ds'].dt.to_pydatetime(), uncertainty_plot_data_valid['yhat_lower'], uncertainty_plot_data_valid['yhat_upper'], color='#0072B2', alpha=0.2, label='Intervalo Incertidumbre')

                # d) Reales Futuros (si existen)
                actual_future_plot_data_valid = comparison_df_plot_full.dropna(subset=['y_future_actual'])
                actual_future_plot_data_valid = actual_future_plot_data_valid[actual_future_plot_data_valid['y_future_actual'] > 0]
                if not actual_future_plot_data_valid.empty: ax_comp.plot(actual_future_plot_data_valid['ds'].dt.to_pydatetime(), actual_future_plot_data_valid['y_future_actual'], 'r.', markersize=5, label='Datos Reales (Posteriores Entreno)')
                # No mostrar advertencia si está vacío, ya es esperado en la primera ejecución

                # e) Línea fin entrenamiento
                ax_comp.axvline(last_train_date_plot, color='grey', linestyle=':', lw=2, label=f'Fin Entrenamiento ({last_train_date_plot.strftime("%Y-%m-%d")})')

                # --- Configuración ---
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2); ax_comp.grid(True, which='minor', c='gray', ls=':', lw=0.5, alpha=0.1)
                ax_comp.set_xlabel('Fecha'); ax_comp.set_ylabel('Valor (adj_close) - Escala Logarítmica')
                ax_comp.set_title(f'Histórico y Predicción Futura para {etf_symbol} (Escala Log)')
                ax_comp.set_yscale('log')
                print("Eje Y configurado a escala logarítmica.")

                # Ajustar límites X (desde inicio histórico hasta fin predicción)
                plot_start_date = comparison_df_plot_full['ds'].min()
                plot_end_date = comparison_df_plot_full['ds'].max()
                if pd.notna(plot_start_date) and pd.notna(plot_end_date):
                    ax_comp.set_xlim([plot_start_date - pd.Timedelta(days=30), plot_end_date + pd.Timedelta(days=30)]) # Margen

                ax_comp.legend(); plt.tight_layout(); plt.show()
                print(f"Gráfico generado para {etf_symbol}.")

            except Exception as e_plot:
                print(f"Error generando gráfico para {etf_symbol}: {e_plot}"); traceback.print_exc()

        elif 'last_real_date' not in results: print(f"Saltando gráfico {etf_symbol}: Falta 'last_real_date'.")
        else: print(f"Saltando gráfico {etf_symbol}: No 'forecast'.")

        # Limpieza post-plot
        if 'forecast_data' in locals(): del forecast_data
        if 'actual_data_etf_plot' in locals(): del actual_data_etf_plot
        if 'historical_data_plot' in locals(): del historical_data_plot
        if 'future_actual_data_plot' in locals(): del future_actual_data_plot
        if 'comparison_df_plot_full' in locals(): del comparison_df_plot_full
        if 'hist_plot_data_valid' in locals(): del hist_plot_data_valid
        if 'forecast_plot_data_valid' in locals(): del forecast_plot_data_valid
        if 'uncertainty_plot_data_valid' in locals(): del uncertainty_plot_data_valid
        if 'actual_future_plot_data_valid' in locals(): del actual_future_plot_data_valid
        if 'fig_comp' in locals(): plt.close(fig_comp)
        gc.collect()

    print("\n--- Fin de la Generación de Gráficos ---")
else:
    print("\nNo se generaron gráficos de comparación (problemas al recargar datos o no hay datos válidos).")

# Limpiar el dataframe recargado
if 'df_actual_new_for_plot' in locals(): del df_actual_new_for_plot; gc.collect()

print("\n--- Script Finalizado ---")
```

```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

*** VERSIÓN ADAPTADA (Basada en Script 1 + Mejoras Script 2) ***

1.  Carga datos históricos.
2.  Selecciona Top N ETFs según NÚMERO DE DATOS.
3.  Opcionalmente aplica Transformación Logarítmica.        ## NUEVO ##
4.  Opcionalmente usa Regresores Externos.                   ## NUEVO ##
5.  Opcionalmente realiza Ajuste de Hiperparámetros (Tuning). ## NUEVO ##
6.  Entrena modelos Prophet (con params base o tuneados).
7.  Genera forecasts (revirtiendo log si aplica).
8.  Muestra resumen de resultados (con info de tuning).
9.  Carga datos reales futuros desde un archivo separado.
10. Genera gráficos comparando predicción vs realidad (EJE Y LOG). ## MODIFICADO ##
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector
import itertools # ## NUEVO ## Para el grid search
import traceback # ## NUEVO ## Para errores detallados

# Instala Prophet si no lo tienes: pip install prophet cmdstanpy
try:
    from prophet import Prophet
    # from prophet.plot import plot_plotly, plot_components_plotly # Descomentar si usas plotly
    # from prophet.plot import add_changepoints_to_plot # No usado en plot final
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30); print("ERROR: 'prophet' no instalado. Ejecuta: pip install prophet cmdstanpy"); print("*"*30); exit()

# Ignorar advertencias
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN GENERAL ---
TOP_N_ETFS = 5
DATASET_VARIABLE_NAME = 'df_top20' # <<< CONFIRMA NOMBRE VARIABLE DATAFRAME ENTRENAMIENTO >>>
FUTURE_END_DATE = '2025-03-10'
ACTUAL_DATA_FILE = 'ETF_cohortes.csv' # <--- NOMBRE DEL ARCHIVO CON DATOS REALES ACTUALIZADOS/COMPLETOS

# --- 1.1 CONFIGURACIÓN DE MEJORAS (Adaptado de Script 2) --- ## NUEVO ## / ## MODIFICADO ##
USE_LOG_TRANSFORM = True             # Poner a True para usar transformación logarítmica y revertirla
REGRESSOR_COLUMNS = []               # Lista de nombres de columnas a usar como regresores (ej: ['VIX_Close', 'SP500_Return'])
PERFORM_HYPERPARAMETER_TUNING = True # Poner a True para buscar los mejores hiperparámetros vía CV
EJECUTAR_CROSS_VALIDATION = False     # NECESARIO para Tuning. Si es False, se omite Tuning y CV.

# --- 1.2 PARÁMETROS BASE Y DE TUNING (Adaptado de Script 2) --- ## NUEVO ## / ## MODIFICADO ##
PROPHET_BASE_PARAMS = {              # Parámetros base que se usarán si no hay tuning o como punto de partida
    'seasonality_mode': 'multiplicative',
    # 'changepoint_prior_scale': 1.0, # Estos ahora están en el grid
    # 'seasonality_prior_scale': 5.0,
    # 'holidays_prior_scale': 5.0,
    'daily_seasonality': False,
    'weekly_seasonality': True,
    'yearly_seasonality': True,
    'interval_width': 0.80
}
CUSTOM_HOLIDAYS = None               # DataFrame de feriados personalizados si se tienen

# --- Parámetros para Grid Search (Solo si PERFORM_HYPERPARAMETER_TUNING es True) ---
PARAM_GRID = {
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [1.0, 10.0, 20.0],
    'holidays_prior_scale': [1.0, 10.0, 20.0] # Asegúrate que CUSTOM_HOLIDAYS esté definido si usas esto
}
CV_METRIC_TO_OPTIMIZE = 'rmse'      # Métrica a minimizar durante el tuning ('mae', 'rmse', 'mape', 'mdape', 'smape', 'coverage')

# --- Parámetros de Cross-Validation (Usados para Tuning o Evaluación separada si Tuning=False) ---
CV_INITIAL = '730 days'
CV_PERIOD = '90 days'
CV_HORIZON = '180 days'


# --- 2. CARGA Y PREPARACIÓN INICIAL DEL DATAFRAME DE ENTRENAMIENTO --- ## MODIFICADO ##

if DATASET_VARIABLE_NAME not in globals():
     raise NameError(f"El DataFrame '{DATASET_VARIABLE_NAME}' no se encontró.")
else:
    df_analysis = globals()[DATASET_VARIABLE_NAME].copy()
    print(f"Usando el DataFrame '{DATASET_VARIABLE_NAME}' para entrenamiento.")

# Verificar columnas requeridas (incluyendo regresores si existen)
required_cols_train = ['fund_symbol', 'price_date', 'adj_close'] + REGRESSOR_COLUMNS # ## MODIFICADO ##
missing_cols = [col for col in required_cols_train if col not in df_analysis.columns]
if missing_cols:
    raise ValueError(f"Faltan columnas requeridas en {DATASET_VARIABLE_NAME}: {missing_cols}")

# Convertir fecha a datetime y manejar errores
try:
    df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date'])
    print("Columna 'price_date' del dataset de entrenamiento convertida a datetime.")
except Exception as e:
    print(f"Error al convertir 'price_date' en {DATASET_VARIABLE_NAME}: {e}"); exit()

# Eliminar filas donde el target es NaN
df_analysis.dropna(subset=['adj_close'], inplace=True)
print(f"Filas con 'adj_close' NaN eliminadas. Datos restantes: {len(df_analysis)}")

# ## NUEVO ## : Pre-procesar NaNs en columnas de regresores (relleno forward/backward por grupo)
if REGRESSOR_COLUMNS:
    print(f"Pre-procesando NaNs en regresores de entrenamiento: {REGRESSOR_COLUMNS}")
    for col in REGRESSOR_COLUMNS:
        original_nan_count = df_analysis[col].isnull().sum()
        if original_nan_count > 0:
            # Rellenar NaNs dentro de cada grupo de ETF
            df_analysis[col] = df_analysis.groupby('fund_symbol')[col].transform(lambda x: x.ffill().bfill())
            remaining_nan_count = df_analysis[col].isnull().sum()
            print(f"  - Columna '{col}': {original_nan_count} NaNs iniciales -> {remaining_nan_count} NaNs restantes.")
            if remaining_nan_count > 0:
                 print(f"  ADVERTENCIA: Aún quedan {remaining_nan_count} NaNs en '{col}' después de ffill/bfill. Podrían causar problemas.")
        # Convertir a numérico por si acaso
        df_analysis[col] = pd.to_numeric(df_analysis[col], errors='coerce')
        if df_analysis[col].isnull().any():
             print(f"  ADVERTENCIA: Columna '{col}' contiene NaNs después de conversión numérica.")


# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---
# (Sin cambios funcionales respecto al Script 1 original, ya hacía esto)
print(f"\nIdentificando los {TOP_N_ETFS} ETFs con mayor número de puntos de datos...")
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()
if not top_N_etfs_list: raise ValueError("No se pudieron contar los datos por ETF o el DataFrame está vacío.")
if len(top_N_etfs_list) < TOP_N_ETFS:
    print(f"Advertencia: Solo se encontraron {len(top_N_etfs_list)} ETFs en el DataFrame.")
    TOP_N_ETFS = len(top_N_etfs_list)
print(f"Se procesarán los siguientes {TOP_N_ETFS} ETFs (seleccionados por tener más datos):")
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
for i, (index, row) in enumerate(selected_counts_info.iterrows()):
    print(f"  {i+1}. {row['fund_symbol']} ({row['data_count']} puntos)")
print("-" * 50)


# --- 4. BUCLE PARA PROCESAR CADA ETF SELECCIONADO ---
results_prophet = {}

for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    # Inicializar resultados para este ETF
    results_prophet[etf_symbol] = {'params': {}, 'log_transform_applied': False} # ## MODIFICADO ##
    is_log_transformed_etf = False # Flag para este ETF específico

    # --- 4.1 Filtrar y Preparar Datos de ENTRENAMIENTO --- ## MODIFICADO ##
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy()
    df_etf = df_etf.sort_values(by='price_date').reset_index(drop=True)

    # Seleccionar columnas necesarias (target + regresores)
    cols_to_keep = ['price_date', 'adj_close'] + REGRESSOR_COLUMNS
    df_prophet_train = df_etf[cols_to_keep].rename(columns={'price_date': 'ds', 'adj_close': 'y'})

    # Eliminar filas con NaNs en 'ds' o 'y' (y en regresores si existen)
    subset_dropna = ['ds', 'y'] + REGRESSOR_COLUMNS
    df_prophet_train.dropna(subset=subset_dropna, inplace=True)

    # --- ## NUEVO ## : Aplicar Transformación Logarítmica (Opcional) ---
    if USE_LOG_TRANSFORM:
        if (df_prophet_train['y'] <= 0).any():
            print(f"ADVERTENCIA: Se encontraron valores de 'y' <= 0 para {etf_symbol}. No se aplicará transformación logarítmica.")
            results_prophet[etf_symbol]['log_transform_applied'] = False
        else:
            df_prophet_train['y'] = np.log(df_prophet_train['y'])
            is_log_transformed_etf = True # Marcar que se usó log para este ETF
            results_prophet[etf_symbol]['log_transform_applied'] = True
            print(f"Transformación Logarítmica aplicada a 'y' para {etf_symbol}.")

    # --- Calcular si hay suficientes datos para CV/Tuning ---
    min_data_points_total = 30 # Mínimo absoluto para entrenar
    can_run_cv_tuning = True  # Asumir que se puede ejecutar CV/Tuning
    try:
        # Verificar mínimo de puntos totales
        if len(df_prophet_train) < min_data_points_total:
            raise ValueError(f"Datos de entrenamiento insuficientes ({len(df_prophet_train)})")

        # Verificar rango de fechas suficiente para la configuración de CV
        min_data_points_cv_days = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days
        data_range_days = (df_prophet_train['ds'].max() - df_prophet_train['ds'].min()).days
        if data_range_days < min_data_points_cv_days:
             print(f"ADVERTENCIA: Rango de datos ({data_range_days} días) insuficiente para CV/Tuning con horizonte {CV_HORIZON}. Se omitirá CV/Tuning.")
             can_run_cv_tuning = False

    except Exception as e_cv_check:
        print(f"ERROR verificando suficiencia de datos para {etf_symbol}: {e_cv_check}")
        results_prophet[etf_symbol]['error'] = f"Error verificando datos: {e_cv_check}"
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect()
        continue # Saltar al siguiente ETF

    last_train_date = df_prophet_train['ds'].max()
    results_prophet[etf_symbol]['last_real_date'] = last_train_date # Guardar última fecha para comparación
    print(f"Datos de entrenamiento listos. Última fecha: {last_train_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet_train)}")

    # --- 4.2 ## NUEVO ## : Ajuste de Hiperparámetros (Tuning) Opcional ---
    best_params_for_etf = PROPHET_BASE_PARAMS.copy() # Empezar con los base
    tuning_ran_successfully = False
    best_cv_metric_value = None

    # Solo ejecutar si está activado, hay suficientes datos, y CV está habilitado globalmente
    if PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION and can_run_cv_tuning:
        print(f"\n--- Iniciando Ajuste de Hiperparámetros (Tuning) para {etf_symbol} ---")
        print(f"Métrica a optimizar: {CV_METRIC_TO_OPTIMIZE}")
        print(f"Grid de parámetros: {PARAM_GRID}")

        # Generar todas las combinaciones de parámetros del grid
        all_params = [dict(zip(PARAM_GRID.keys(), v)) for v in itertools.product(*PARAM_GRID.values())]
        tuning_results = [] # Lista para guardar {'params': ..., 'metric': ...}

        print(f"Probando {len(all_params)} combinaciones de parámetros...")
        for i, params_to_try in enumerate(all_params):
            current_trial_params = {**PROPHET_BASE_PARAMS, **params_to_try} # Combinar base + grid
            print(f"  Prueba {i+1}/{len(all_params)}: {params_to_try}", end="")
            try:
                # Instanciar modelo temporal para CV
                m_tune = Prophet(**current_trial_params, holidays=CUSTOM_HOLIDAYS)
                # Añadir regresores si existen
                if REGRESSOR_COLUMNS:
                    for regressor in REGRESSOR_COLUMNS:
                        m_tune.add_regressor(regressor)
                # Entrenar modelo temporal
                m_tune.fit(df_prophet_train)
                # Ejecutar Cross-Validation
                df_cv_tune = cross_validation(m_tune, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes", disable_tqdm=True)
                # Calcular métricas de rendimiento
                df_p_tune = performance_metrics(df_cv_tune)
                # Obtener el valor promedio de la métrica objetivo
                metric_value = df_p_tune[CV_METRIC_TO_OPTIMIZE].mean()
                tuning_results.append({'params': params_to_try, 'metric': metric_value})
                print(f" -> {CV_METRIC_TO_OPTIMIZE}={metric_value:.4f}")
                # Limpiar memoria del modelo temporal
                del m_tune, df_cv_tune, df_p_tune; gc.collect()

            except Exception as e_tune:
                print(f" -> ERROR: {e_tune}")
                # Guardar resultado como infinito para que no sea elegido como el mejor
                tuning_results.append({'params': params_to_try, 'metric': float('inf')})
                # Limpieza adicional en caso de error
                if 'm_tune' in locals(): del m_tune
                if 'df_cv_tune' in locals(): del df_cv_tune
                if 'df_p_tune' in locals(): del df_p_tune
                gc.collect()

        # Encontrar la mejor combinación (menor métrica)
        if tuning_results:
            best_run = min(tuning_results, key=lambda x: x['metric'])
            # Asegurarse que el mejor run no fue un error (inf)
            if best_run['metric'] != float('inf'):
                best_params_for_etf.update(best_run['params']) # Actualizar los params base con los mejores del grid
                best_cv_metric_value = best_run['metric']
                print(f"\nMejores parámetros encontrados para {etf_symbol}: {best_run['params']}")
                print(f"Mejor valor de {CV_METRIC_TO_OPTIMIZE} (promedio CV): {best_cv_metric_value:.4f}")
                tuning_ran_successfully = True
            else:
                print("\nEl Tuning falló para todas las combinaciones. Se usarán los parámetros base.")
        else:
            print("\nNo se obtuvieron resultados del Tuning. Se usarán los parámetros base.")

    else:
        # Mensajes indicando por qué no se hizo tuning
        if not PERFORM_HYPERPARAMETER_TUNING: print("\nAjuste de Hiperparámetros OMITIDO (desactivado por configuración).")
        elif not EJECUTAR_CROSS_VALIDATION: print("\nAjuste de Hiperparámetros OMITIDO (Cross-Validation global desactivada).")
        elif not can_run_cv_tuning: print("\nAjuste de Hiperparámetros OMITIDO (datos insuficientes para la configuración de CV).")

    # Guardar los parámetros finales usados (base o tuneados)
    results_prophet[etf_symbol]['params'] = best_params_for_etf.copy()
    print(f"Parámetros finales a usar para el modelo de {etf_symbol}: {best_params_for_etf}")


    # --- 4.3 Entrenar Modelo FINAL --- ## MODIFICADO ##
    print(f"\nEntrenando modelo FINAL para {etf_symbol}...")
    model_final = Prophet(**best_params_for_etf, holidays=CUSTOM_HOLIDAYS)

    # Añadir regresores si se especificaron
    if REGRESSOR_COLUMNS:
        print(f"Añadiendo regresores al modelo final: {REGRESSOR_COLUMNS}")
        for regressor in REGRESSOR_COLUMNS:
            model_final.add_regressor(regressor)

    # Añadir feriados de USA (si no se usan CUSTOM_HOLIDAYS que los incluyan)
    # Descomentar si es necesario y CUSTOM_HOLIDAYS es None
    # try:
    #     if CUSTOM_HOLIDAYS is None:
    #         model_final.add_country_holidays(country_name='US')
    #         print("Feriados de USA añadidos al modelo final.")
    # except Exception as e_hol:
    #     print(f"Advertencia: No se pudieron añadir feriados de USA: {e_hol}")

    try:
        model_final.fit(df_prophet_train)
        print("Entrenamiento del modelo final completado.")
    except Exception as e_fit_final:
        print(f"ERROR FATAL al entrenar el modelo FINAL para {etf_symbol}: {e_fit_final}")
        traceback.print_exc() # Imprimir detalles del error
        results_prophet[etf_symbol]['error'] = f"Error entrenamiento final: {e_fit_final}"
        # Limpieza robusta en caso de fallo
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect()
        continue # Saltar al siguiente ETF

    # --- 4.4 Resumen CV/Tuning --- ## MODIFICADO ##
    # Ya no ejecutamos CV separada aquí. Guardamos un resumen del tuning.
    cv_summary_message = "CV/Tuning no ejecutado o fallido."
    if tuning_ran_successfully:
        cv_summary_message = f"Tuning OK. Mejor {CV_METRIC_TO_OPTIMIZE}: {best_cv_metric_value:.4f}"
    elif not PERFORM_HYPERPARAMETER_TUNING:
        cv_summary_message = "Tuning desactivado."
    elif not EJECUTAR_CROSS_VALIDATION:
         cv_summary_message = "Tuning omitido (CV global desactivada)."
    elif not can_run_cv_tuning:
         cv_summary_message = "Tuning omitido (datos insuficientes)."

    results_prophet[etf_symbol]['cv_metrics_summary'] = cv_summary_message # Guardar el mensaje resumen


    # --- 4.5 Generar Forecast Final --- ## MODIFICADO ##
    print(f"\nGenerando Forecast Final para {etf_symbol} hasta {FUTURE_END_DATE}...")
    forecast_final_scale = None # Variable para el forecast en escala original
    future_df_final = None # Variable para el dataframe futuro

    try:
        # Crear dataframe futuro base con las fechas
        target_end_date = pd.to_datetime(FUTURE_END_DATE)
        freq = 'B' # Frecuencia de días hábiles (Business days)
        # Asegurar que empezamos un día después del último dato de entrenamiento
        future_dates = pd.date_range(start=last_train_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)

        if future_dates.empty:
             raise ValueError(f"No se generaron fechas futuras (quizás {FUTURE_END_DATE} es anterior a {last_train_date.strftime('%Y-%m-%d')})")

        future_df_final = pd.DataFrame({'ds': future_dates})
        future_df_final = future_df_final[future_df_final['ds'] <= target_end_date] # Asegurar no pasar la fecha final

        # --- ## NUEVO ## : Añadir valores futuros para los regresores ---
        if REGRESSOR_COLUMNS:
            print(f"Preparando valores futuros para regresores: {REGRESSOR_COLUMNS}")
            try:
                # Intentar cargar datos completos que podrían tener los regresores futuros
                df_actual_full = pd.read_csv(ACTUAL_DATA_FILE)
                df_actual_full['ds'] = pd.to_datetime(df_actual_full['price_date']) # Usar 'ds' temporalmente
                df_reg_future_values = df_actual_full.loc[df_actual_full['fund_symbol'] == etf_symbol, ['ds'] + REGRESSOR_COLUMNS].copy()

                # Hacer merge con el dataframe futuro
                future_df_final = pd.merge(future_df_final, df_reg_future_values, on='ds', how='left')

                # Rellenar posibles NaNs en los regresores futuros
                for col in REGRESSOR_COLUMNS:
                    # Primero, ffill (usar el último valor conocido)
                    future_df_final[col] = future_df_final[col].ffill()
                    # Si aún quedan NaNs (al principio), usar el último valor del set de entrenamiento
                    if future_df_final[col].isnull().any():
                        last_known_regressor_value = df_prophet_train[col].iloc[-1]
                        future_df_final[col].fillna(last_known_regressor_value, inplace=True)
                        print(f"  ADVERTENCIA: NaNs en regresor futuro '{col}' rellenados con último valor conocido ({last_known_regressor_value:.4f}).")

                    # Verificar si aún quedan NaNs (no debería pasar, pero por seguridad)
                    if future_df_final[col].isnull().any():
                         raise ValueError(f"No se pudieron obtener/rellenar valores futuros para el regresor '{col}'.")

                del df_actual_full, df_reg_future_values # Limpiar memoria
                print("Valores futuros de regresores añadidos.")

            except FileNotFoundError:
                 raise ValueError(f"No se encontró el archivo '{ACTUAL_DATA_FILE}' necesario para obtener regresores futuros.")
            except Exception as e_reg_fut:
                 raise ValueError(f"Error al preparar regresores futuros: {e_reg_fut}")

        # Realizar la predicción con el modelo final
        print(f"Realizando predicción con {len(future_df_final)} fechas futuras...")
        if future_df_final.empty:
             raise ValueError("El DataFrame futuro está vacío antes de la predicción.")

        forecast_raw = model_final.predict(future_df_final)
        print("Predicción completada.")

        # Copiar el forecast para trabajar sobre él
        forecast_final_scale = forecast_raw.copy()

        # --- ## NUEVO ## : Revertir Transformación Logarítmica si se aplicó ---
        if is_log_transformed_etf:
            print(f"Revirtiendo transformación logarítmica para {etf_symbol}...")
            cols_to_exp = ['yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']
            # Incluir componentes de estacionalidad si existen en el forecast_raw
            for col in forecast_raw.columns:
                 if 'seasonality' in col or col in ('weekly', 'yearly', 'daily', 'multiplicative_terms', 'additive_terms'): # Añadir otros componentes si es necesario
                     if col not in cols_to_exp:
                         cols_to_exp.append(col)
                 # Incluir regresores si existen
                 if col in REGRESSOR_COLUMNS and col + '_lower' not in cols_to_exp: # Revertir efecto del regresor
                    cols_to_exp.extend([col, col + '_lower', col + '_upper'])


            cols_reverted = []
            for col in cols_to_exp:
                if col in forecast_final_scale.columns:
                    try:
                        # Evitar revertir valores extremadamente bajos que puedan dar inf o NaN
                        # forecast_final_scale[col] = np.exp(np.clip(forecast_final_scale[col], a_min=-700, a_max=700)) # Clip para evitar overflow
                        forecast_final_scale[col] = np.exp(forecast_final_scale[col])
                        cols_reverted.append(col)
                    except Exception as e_exp:
                        print(f"  ERROR al revertir (np.exp) columna '{col}': {e_exp}. Dejando como está.")
                        # Podrías optar por poner NaN si falla la reversión: forecast_final_scale[col] = np.nan
            print(f"Columnas revertidas a escala original: {cols_reverted}")

        # Calcular ancho del intervalo de predicción en escala final
        if 'yhat_upper' in forecast_final_scale.columns and 'yhat_lower' in forecast_final_scale.columns:
             forecast_final_scale['prediction_interval_width'] = forecast_final_scale['yhat_upper'] - forecast_final_scale['yhat_lower']

        # Guardar el forecast relevante (en escala original)
        cols_to_save = ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']
        cols_present_in_forecast = [c for c in cols_to_save if c in forecast_final_scale.columns]
        results_prophet[etf_symbol]['forecast'] = forecast_final_scale[cols_present_in_forecast].copy()

        print(f"Forecast final (escala original) generado y guardado. Última fecha: {forecast_final_scale['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e_pred:
        print(f"ERROR FATAL generando el forecast final para {etf_symbol}: {e_pred}")
        traceback.print_exc() # Imprimir detalles
        results_prophet[etf_symbol]['error_forecast'] = str(e_pred)
        # Limpieza robusta en caso de fallo en predicción
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        if 'future_df_final' in locals(): del future_df_final
        if 'forecast_raw' in locals(): del forecast_raw
        if 'forecast_final_scale' in locals(): del forecast_final_scale
        gc.collect()
        continue # Saltar al siguiente ETF


    # --- 4.6 Visualización Inicial (Opcional - Comentada por defecto) ---
    # print(f"\nGenerando gráficos iniciales para {etf_symbol}...")
    # try:
    #     if forecast_final_scale is not None: # Usar el forecast en escala original
    #         # Nota: model_final.plot() usará la escala en la que fue entrenado (log si aplica).
    #         # Para plotear en escala original, puede ser mejor hacerlo manualmente o ajustar ejes.
    #         fig1 = model_final.plot(forecast_raw) # Plotea sobre el forecast antes de revertir log
    #         plt.axvline(last_train_date, color='gray', linestyle=':', lw=2, label='Último Dato Entrenamiento')
    #         y_label = "Valor (y)" if not is_log_transformed_etf else "Valor (log(y))"
    #         plt.title(f'Predicción Prophet Base para {etf_symbol}')
    #         plt.xlabel("Fecha"); plt.ylabel(y_label)
    #         plt.legend(); plt.tight_layout(); plt.show()

    #         fig2 = model_final.plot_components(forecast_raw) # Componentes también en escala de entrenamiento
    #         plt.suptitle(f'Componentes del Modelo Prophet para {etf_symbol}', y=1.02)
    #         plt.tight_layout(); plt.show()
    #         print(f"Gráficos iniciales generados.")
    #     else:
    #         print("No se generó forecast, omitiendo gráficos iniciales.")
    # except Exception as e_plot_init:
    #     print(f"Error generando gráficos iniciales: {e_plot_init}")

    # --- 4.7 Limpieza de Memoria --- ## MODIFICADO ## (Más completo)
    print(f"Limpiando memoria para {etf_symbol}...")
    del model_final, df_etf, df_prophet_train
    if 'future_df_final' in locals(): del future_df_final
    if 'forecast_raw' in locals(): del forecast_raw
    if 'forecast_final_scale' in locals(): del forecast_final_scale
    # Variables de tuning
    if 'tuning_results' in locals(): del tuning_results
    if 'best_run' in locals(): del best_run
    if 'all_params' in locals(): del all_params
    # Variables de CV si se usaron en tuning
    # No es necesario borrar m_tune, df_cv_tune, df_p_tune aquí porque se borran dentro del bucle de tuning
    gc.collect()
    print("Limpieza completada.")

    print(f"\n{'='*15} Procesamiento de {etf_symbol} completado {'='*15}")
    print("-" * 70)


print("\n###########################################################")
print("   ANÁLISIS COMPLETADO PARA TODOS LOS ETFs SELECCIONADOS")
print("###########################################################")

# --- 5. MOSTRAR RESUMEN FINAL DE RESULTADOS DEL ANÁLISIS --- ## MODIFICADO ## (para mostrar resumen CV/Tuning)
print("\n--- Resumen de Resultados del Análisis ---")
for etf, data in results_prophet.items():
    print(f"\n--- ETF: {etf} ---")
    print(f"  Transformación Log Aplicada: {data.get('log_transform_applied', 'N/D')}")
    print(f"  Parámetros Finales Usados: {data.get('params', 'N/A')}")
    if 'error' in data:
        print(f"  ERROR PRINCIPAL: {data['error']}")
        continue # No mostrar más si hubo error temprano

    # Mostrar resumen de CV/Tuning
    print("\n  Resultado Cross-Validation / Tuning:")
    cv_summary = data.get('cv_metrics_summary', 'No disponible')
    print(f"    {cv_summary}") # Mostrar el string resumen

    # Mostrar ejemplo de predicciones finales
    print("\n  Ejemplo Predicciones Finales (Últimas 5 - Escala Original):")
    if 'forecast' in data and not data['forecast'].empty:
        # Asegurarse que las columnas existen antes de intentar mostrarlas
        cols_to_display = [c for c in ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width'] if c in data['forecast'].columns]
        if cols_to_display:
             final_preds_display = data['forecast'][cols_to_display].tail()
             print(final_preds_display.round(4).to_string(index=False))
        else:
             print("    No se encontraron columnas estándar de forecast para mostrar.")
    elif 'error_forecast' in data:
         print(f"    ERROR en Forecast: {data['error_forecast']}")
    else:
         print("    No se encontraron datos de forecast.")
    print("-" * 40)
print("\n--- Fin del Resumen del Análisis ---")


# --- 6. COMPARACIÓN CON DATOS REALES ACTUALIZADOS ---
# (Carga y preparación inicial sin cambios respecto a Script 1)

print("\n\n###########################################################")
print(f"      COMPARACIÓN CON DATOS REALES DESDE '{ACTUAL_DATA_FILE}'")
print("###########################################################")

df_actual_new = None # Inicializar
comparison_possible = False
try:
    # Cargar los datos reales completos/actualizados
    df_actual_new = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"Datos reales cargados desde '{ACTUAL_DATA_FILE}'.")

    # --- Preparación de Datos Reales ---
    DATE_COL_ACTUAL = 'price_date'
    VALUE_COL_ACTUAL = 'adj_close'
    SYMBOL_COL_ACTUAL = 'fund_symbol'
    required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new.columns for col in required_actual_cols):
        raise ValueError(f"'{ACTUAL_DATA_FILE}' debe contener: {required_actual_cols}")

    df_actual_new[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new[DATE_COL_ACTUAL])
    df_actual_new = df_actual_new.rename(columns={
        DATE_COL_ACTUAL: 'ds',
        VALUE_COL_ACTUAL: 'y_actual',
        SYMBOL_COL_ACTUAL: 'fund_symbol'
    })
    # Eliminar NaNs en columnas clave para la comparación
    df_actual_new.dropna(subset=['ds', 'y_actual', 'fund_symbol'], inplace=True)
    df_actual_new = df_actual_new.sort_values(by=['fund_symbol', 'ds'])
    print("Datos reales preparados (columnas renombradas, NaNs eliminados).")
    comparison_possible = True

except FileNotFoundError:
    print(f"ERROR: El archivo '{ACTUAL_DATA_FILE}' no se encontró.")
    print("Asegúrate de que el archivo exista y la ruta sea correcta.")
    comparison_possible = False
except Exception as e:
    print(f"Error al cargar o preparar '{ACTUAL_DATA_FILE}': {e}")
    traceback.print_exc()
    comparison_possible = False

# --- Generar Gráficos de Comparación (MANUALMENTE con EJE Y LOG) --- ## MODIFICADO ##
if comparison_possible and df_actual_new is not None:
    print("\n--- Generando Gráficos de Comparación (Predicción vs Realidad Futura - EJE Y LOG) ---") # Título actualizado

    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Comparando para ETF: {etf_symbol} ---")

        # Verificar que tenemos forecast y la última fecha de entrenamiento para este ETF
        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data = results['forecast'].copy()
            last_train_date_plot = results['last_real_date'] # Renombrado para claridad en plot

            # Filtrar datos reales actualizados para este ETF y para fechas POSTERIORES al entrenamiento
            actual_data_etf_plot = df_actual_new[
                (df_actual_new['fund_symbol'] == etf_symbol) &
                (df_actual_new['ds'] > last_train_date_plot)
            ].copy()

            if actual_data_etf_plot.empty:
                print(f"No se encontraron datos reales en '{ACTUAL_DATA_FILE}' para {etf_symbol} después de {last_train_date_plot.strftime('%Y-%m-%d')}.")
                continue

            print(f"Encontrados {len(actual_data_etf_plot)} puntos reales futuros para comparar.")

            # --- Creación del Gráfico Manual ---
            try:
                print(f"Preparando gráfico manual para {etf_symbol}")
                # Recuperar datos históricos 'y' del df de análisis original para plotearlos
                # Renombrar a 'y_hist_orig' para evitar colisión con 'y' (log) si se usó
                historical_data_plot = df_analysis.loc[
                    df_analysis['fund_symbol'] == etf_symbol, ['price_date', 'adj_close']
                ].rename(columns={'price_date': 'ds', 'adj_close': 'y_hist_orig'}).dropna()

                # Asegurarse que las columnas necesarias están en el forecast
                required_forecast_cols = ['ds', 'yhat', 'yhat_lower', 'yhat_upper']
                if not all(col in forecast_data.columns for col in required_forecast_cols):
                     print(f"ADVERTENCIA: Faltan columnas en el forecast para {etf_symbol}. Columnas presentes: {forecast_data.columns.tolist()}")
                     # Crear columnas dummy si faltan para evitar error en merge, aunque el plot será incompleto
                     for col in required_forecast_cols:
                         if col not in forecast_data.columns: forecast_data[col] = np.nan


                # Unir: forecast (predicciones futuras) + datos reales futuros + datos históricos
                # 1. Merge forecast con datos reales futuros
                comparison_df_plot = pd.merge(forecast_data[required_forecast_cols],
                                              actual_data_etf_plot[['ds', 'y_actual']],
                                              on='ds', how='left')
                # 2. Merge resultado con datos históricos
                # Usar outer merge para incluir todo el rango de fechas (histórico + futuro)
                comparison_df_plot_full = pd.merge(historical_data_plot[['ds', 'y_hist_orig']],
                                                   comparison_df_plot,
                                                   on='ds', how='outer')

                # Ordenar por fecha para el gráfico
                comparison_df_plot_full = comparison_df_plot_full.sort_values(by='ds').reset_index(drop=True)
                print(f"DataFrame de comparación preparado con {len(comparison_df_plot_full)} filas.")

                # Crear figura y ejes
                fig_comp, ax_comp = plt.subplots(figsize=(14, 7)) # Tamaño un poco más grande

                # --- Graficar elementos ---

                # a) Datos históricos observados (puntos negros) - Filtrar > 0 para log
                hist_plot_data = comparison_df_plot_full.dropna(subset=['y_hist_orig'])
                hist_plot_data = hist_plot_data[hist_plot_data['y_hist_orig'] > 0]
                if not hist_plot_data.empty:
                    ax_comp.plot(hist_plot_data['ds'].dt.to_pydatetime(), hist_plot_data['y_hist_orig'],
                                 'k.', markersize=3, label='Datos Observados (Históricos)')

                # b) Predicción (línea azul) - Filtrar > 0 para log
                forecast_plot_data = comparison_df_plot_full.dropna(subset=['yhat'])
                forecast_plot_data = forecast_plot_data[forecast_plot_data['yhat'] > 0]
                if not forecast_plot_data.empty:
                    ax_comp.plot(forecast_plot_data['ds'].dt.to_pydatetime(), forecast_plot_data['yhat'],
                                 ls='-', c='#0072B2', label='Predicción (yhat)')

                # c) Intervalo de incertidumbre (área sombreada azul) - Filtrar > 0 para log
                uncertainty_plot_data = comparison_df_plot_full.dropna(subset=['yhat_lower', 'yhat_upper'])
                uncertainty_plot_data = uncertainty_plot_data[(uncertainty_plot_data['yhat_lower'] > 0) & (uncertainty_plot_data['yhat_upper'] > 0)]
                if not uncertainty_plot_data.empty:
                     ax_comp.fill_between(uncertainty_plot_data['ds'].dt.to_pydatetime(),
                                          uncertainty_plot_data['yhat_lower'], uncertainty_plot_data['yhat_upper'],
                                          color='#0072B2', alpha=0.2, label='Intervalo Incertidumbre')

                # d) Datos reales futuros (puntos rojos) - Filtrar > 0 para log
                actual_plot_data = comparison_df_plot_full.dropna(subset=['y_actual'])
                actual_plot_data = actual_plot_data[actual_plot_data['y_actual'] > 0]
                if not actual_plot_data.empty:
                    ax_comp.plot(actual_plot_data['ds'].dt.to_pydatetime(), actual_plot_data['y_actual'],
                                 'r.', markersize=5, label='Datos Reales (Futuros)')
                else:
                     print(f"Advertencia: No hay datos 'y_actual' positivos para graficar en escala log.")

                # e) Línea vertical indicando fin del entrenamiento
                ax_comp.axvline(last_train_date_plot, color='grey', linestyle=':', lw=2, label='Fin Entrenamiento')

                # --- Configuración del Gráfico ---
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
                ax_comp.grid(True, which='minor', c='gray', ls=':', lw=0.5, alpha=0.1) # Añadir grid menor
                ax_comp.set_xlabel('Fecha')
                # ## MODIFICADO ## : Etiqueta y escala del eje Y
                ax_comp.set_ylabel('Valor (adj_close) - Escala Logarítmica')
                ax_comp.set_title(f'Comparación Predicción vs Realidad para {etf_symbol} (Escala Log)')
                ax_comp.set_yscale('log') # <-- Establecer escala logarítmica
                print("Eje Y configurado a escala logarítmica.")

                # Ajustar límites del eje X para enfocar en el período de interés
                # Mostrar un poco antes del fin del entrenamiento y hasta el final del forecast/realidad
                days_before = 60 # Mostrar ~2 meses antes
                plot_start_date = last_train_date_plot - pd.Timedelta(days=days_before)
                plot_end_date = comparison_df_plot_full['ds'].max()
                # Añadir un pequeño margen al final
                if pd.notna(plot_end_date):
                    ax_comp.set_xlim([plot_start_date, plot_end_date + pd.Timedelta(days=10)])
                else: # Si no hay fechas futuras, usar la fecha de entrenamiento como final
                     ax_comp.set_xlim([plot_start_date, last_train_date_plot + pd.Timedelta(days=10)])


                # Añadir leyenda
                ax_comp.legend()

                # Mostrar
                plt.tight_layout()
                plt.show()
                print(f"Gráfico de comparación (log) generado manualmente para {etf_symbol}.")

            except Exception as e_plot:
                print(f"Error generando el gráfico de comparación manual para {etf_symbol}: {e_plot}")
                traceback.print_exc() # Mostrar error detallado

        elif 'last_real_date' not in results:
             print(f"Saltando gráfico para {etf_symbol}: Falta 'last_real_date' en los resultados.")
        else:
             print(f"Saltando gráfico para {etf_symbol}: No hay 'forecast' disponible en los resultados.")

        # Limpieza robusta después de cada gráfico ## MODIFICADO ##
        if 'forecast_data' in locals(): del forecast_data
        if 'actual_data_etf_plot' in locals(): del actual_data_etf_plot
        if 'historical_data_plot' in locals(): del historical_data_plot
        if 'comparison_df_plot' in locals(): del comparison_df_plot
        if 'comparison_df_plot_full' in locals(): del comparison_df_plot_full
        if 'hist_plot_data' in locals(): del hist_plot_data
        if 'forecast_plot_data' in locals(): del forecast_plot_data
        if 'uncertainty_plot_data' in locals(): del uncertainty_plot_data
        if 'actual_plot_data' in locals(): del actual_plot_data
        if 'fig_comp' in locals(): plt.close(fig_comp) # Cerrar figura explícitamente
        gc.collect()

    print("\n--- Fin de la Comparación ---")
else:
    print("\nNo se generaron gráficos de comparación debido a problemas al cargar o procesar los datos reales.")

print("\n--- Script Finalizado ---")
```

```python
# -*- coding: utf-8 -*-
"""
Script Completo para Análisis y Comparación de ETFs con Prophet.

*** VERSIÓN ADAPTADA (Basada en Script 1 + Mejoras Script 2) ***

1.  Carga datos históricos.
2.  Selecciona Top N ETFs según NÚMERO DE DATOS.
3.  Opcionalmente aplica Transformación Logarítmica.        ## NUEVO ##
4.  Opcionalmente usa Regresores Externos.                   ## NUEVO ##
5.  Opcionalmente realiza Ajuste de Hiperparámetros (Tuning). ## NUEVO ##
6.  Entrena modelos Prophet (con params base o tuneados).
7.  Genera forecasts (revirtiendo log si aplica).
8.  Muestra resumen de resultados (con info de tuning).
9.  Carga datos reales futuros desde un archivo separado.
10. Genera gráficos comparando predicción vs realidad (EJE Y LOG). ## MODIFICADO ##
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import gc # Garbage collector
import itertools # ## NUEVO ## Para el grid search
import traceback # ## NUEVO ## Para errores detallados

# Instala Prophet si no lo tienes: pip install prophet cmdstanpy
try:
    from prophet import Prophet
    # from prophet.plot import plot_plotly, plot_components_plotly # Descomentar si usas plotly
    # from prophet.plot import add_changepoints_to_plot # No usado en plot final
    from prophet.diagnostics import cross_validation, performance_metrics
except ImportError:
    print("*"*30); print("ERROR: 'prophet' no instalado. Ejecuta: pip install prophet cmdstanpy"); print("*"*30); exit()

# Ignorar advertencias
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)


# --- 1. CONFIGURACIÓN GENERAL ---
TOP_N_ETFS = 5
DATASET_VARIABLE_NAME = 'df_top20' # <<< CONFIRMA NOMBRE VARIABLE DATAFRAME ENTRENAMIENTO >>>
FUTURE_END_DATE = '2025-03-10'
ACTUAL_DATA_FILE = 'ETF_cohortes.csv' # <--- NOMBRE DEL ARCHIVO CON DATOS REALES ACTUALIZADOS/COMPLETOS

# --- 1.1 CONFIGURACIÓN DE MEJORAS (Adaptado de Script 2) --- ## NUEVO ## / ## MODIFICADO ##
USE_LOG_TRANSFORM = True             # Poner a True para usar transformación logarítmica y revertirla
REGRESSOR_COLUMNS = []               # Lista de nombres de columnas a usar como regresores (ej: ['VIX_Close', 'SP500_Return'])
PERFORM_HYPERPARAMETER_TUNING = True # Poner a True para buscar los mejores hiperparámetros vía CV
EJECUTAR_CROSS_VALIDATION = True     # NECESARIO para Tuning. Si es False, se omite Tuning y CV.

# --- 1.2 PARÁMETROS BASE Y DE TUNING (Adaptado de Script 2) --- ## NUEVO ## / ## MODIFICADO ##
PROPHET_BASE_PARAMS = {              # Parámetros base que se usarán si no hay tuning o como punto de partida
    'seasonality_mode': 'multiplicative',
    # 'changepoint_prior_scale': 1.0, # Estos ahora están en el grid
    # 'seasonality_prior_scale': 5.0,
    # 'holidays_prior_scale': 5.0,
    'daily_seasonality': False,
    'weekly_seasonality': True,
    'yearly_seasonality': True,
    'interval_width': 0.80
}
CUSTOM_HOLIDAYS = None               # DataFrame de feriados personalizados si se tienen

# --- Parámetros para Grid Search (Solo si PERFORM_HYPERPARAMETER_TUNING es True) ---
PARAM_GRID = {
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [1.0, 10.0, 20.0],
    'holidays_prior_scale': [1.0, 10.0, 20.0] # Asegúrate que CUSTOM_HOLIDAYS esté definido si usas esto
}
CV_METRIC_TO_OPTIMIZE = 'rmse'      # Métrica a minimizar durante el tuning ('mae', 'rmse', 'mape', 'mdape', 'smape', 'coverage')

# --- Parámetros de Cross-Validation (Usados para Tuning o Evaluación separada si Tuning=False) ---
CV_INITIAL = '730 days'
CV_PERIOD = '90 days'
CV_HORIZON = '180 days'


# --- 2. CARGA Y PREPARACIÓN INICIAL DEL DATAFRAME DE ENTRENAMIENTO --- ## MODIFICADO ##

if DATASET_VARIABLE_NAME not in globals():
     raise NameError(f"El DataFrame '{DATASET_VARIABLE_NAME}' no se encontró.")
else:
    df_analysis = globals()[DATASET_VARIABLE_NAME].copy()
    print(f"Usando el DataFrame '{DATASET_VARIABLE_NAME}' para entrenamiento.")

# Verificar columnas requeridas (incluyendo regresores si existen)
required_cols_train = ['fund_symbol', 'price_date', 'adj_close'] + REGRESSOR_COLUMNS # ## MODIFICADO ##
missing_cols = [col for col in required_cols_train if col not in df_analysis.columns]
if missing_cols:
    raise ValueError(f"Faltan columnas requeridas en {DATASET_VARIABLE_NAME}: {missing_cols}")

# Convertir fecha a datetime y manejar errores
try:
    df_analysis['price_date'] = pd.to_datetime(df_analysis['price_date'])
    print("Columna 'price_date' del dataset de entrenamiento convertida a datetime.")
except Exception as e:
    print(f"Error al convertir 'price_date' en {DATASET_VARIABLE_NAME}: {e}"); exit()

# Eliminar filas donde el target es NaN
df_analysis.dropna(subset=['adj_close'], inplace=True)
print(f"Filas con 'adj_close' NaN eliminadas. Datos restantes: {len(df_analysis)}")

# ## NUEVO ## : Pre-procesar NaNs en columnas de regresores (relleno forward/backward por grupo)
if REGRESSOR_COLUMNS:
    print(f"Pre-procesando NaNs en regresores de entrenamiento: {REGRESSOR_COLUMNS}")
    for col in REGRESSOR_COLUMNS:
        original_nan_count = df_analysis[col].isnull().sum()
        if original_nan_count > 0:
            # Rellenar NaNs dentro de cada grupo de ETF
            df_analysis[col] = df_analysis.groupby('fund_symbol')[col].transform(lambda x: x.ffill().bfill())
            remaining_nan_count = df_analysis[col].isnull().sum()
            print(f"  - Columna '{col}': {original_nan_count} NaNs iniciales -> {remaining_nan_count} NaNs restantes.")
            if remaining_nan_count > 0:
                 print(f"  ADVERTENCIA: Aún quedan {remaining_nan_count} NaNs en '{col}' después de ffill/bfill. Podrían causar problemas.")
        # Convertir a numérico por si acaso
        df_analysis[col] = pd.to_numeric(df_analysis[col], errors='coerce')
        if df_analysis[col].isnull().any():
             print(f"  ADVERTENCIA: Columna '{col}' contiene NaNs después de conversión numérica.")


# --- 3. SELECCIONAR TOP N ETFs POR NÚMERO DE DATOS ---
# (Sin cambios funcionales respecto al Script 1 original, ya hacía esto)
print(f"\nIdentificando los {TOP_N_ETFS} ETFs con mayor número de puntos de datos...")
etf_data_counts = df_analysis.groupby('fund_symbol').size().reset_index(name='data_count')
etf_data_counts_sorted = etf_data_counts.sort_values(by='data_count', ascending=False)
top_N_etfs_list = etf_data_counts_sorted['fund_symbol'].head(TOP_N_ETFS).tolist()
if not top_N_etfs_list: raise ValueError("No se pudieron contar los datos por ETF o el DataFrame está vacío.")
if len(top_N_etfs_list) < TOP_N_ETFS:
    print(f"Advertencia: Solo se encontraron {len(top_N_etfs_list)} ETFs en el DataFrame.")
    TOP_N_ETFS = len(top_N_etfs_list)
print(f"Se procesarán los siguientes {TOP_N_ETFS} ETFs (seleccionados por tener más datos):")
selected_counts_info = etf_data_counts_sorted.head(TOP_N_ETFS)
for i, (index, row) in enumerate(selected_counts_info.iterrows()):
    print(f"  {i+1}. {row['fund_symbol']} ({row['data_count']} puntos)")
print("-" * 50)


# --- 4. BUCLE PARA PROCESAR CADA ETF SELECCIONADO ---
results_prophet = {}

for etf_symbol in top_N_etfs_list:
    print(f"\n{'='*15} Procesando ETF: {etf_symbol} {'='*15}")
    # Inicializar resultados para este ETF
    results_prophet[etf_symbol] = {'params': {}, 'log_transform_applied': False} # ## MODIFICADO ##
    is_log_transformed_etf = False # Flag para este ETF específico

    # --- 4.1 Filtrar y Preparar Datos de ENTRENAMIENTO --- ## MODIFICADO ##
    df_etf = df_analysis[df_analysis['fund_symbol'] == etf_symbol].copy()
    df_etf = df_etf.sort_values(by='price_date').reset_index(drop=True)

    # Seleccionar columnas necesarias (target + regresores)
    cols_to_keep = ['price_date', 'adj_close'] + REGRESSOR_COLUMNS
    df_prophet_train = df_etf[cols_to_keep].rename(columns={'price_date': 'ds', 'adj_close': 'y'})

    # Eliminar filas con NaNs en 'ds' o 'y' (y en regresores si existen)
    subset_dropna = ['ds', 'y'] + REGRESSOR_COLUMNS
    df_prophet_train.dropna(subset=subset_dropna, inplace=True)

    # --- ## NUEVO ## : Aplicar Transformación Logarítmica (Opcional) ---
    if USE_LOG_TRANSFORM:
        if (df_prophet_train['y'] <= 0).any():
            print(f"ADVERTENCIA: Se encontraron valores de 'y' <= 0 para {etf_symbol}. No se aplicará transformación logarítmica.")
            results_prophet[etf_symbol]['log_transform_applied'] = False
        else:
            df_prophet_train['y'] = np.log(df_prophet_train['y'])
            is_log_transformed_etf = True # Marcar que se usó log para este ETF
            results_prophet[etf_symbol]['log_transform_applied'] = True
            print(f"Transformación Logarítmica aplicada a 'y' para {etf_symbol}.")

    # --- Calcular si hay suficientes datos para CV/Tuning ---
    min_data_points_total = 30 # Mínimo absoluto para entrenar
    can_run_cv_tuning = True  # Asumir que se puede ejecutar CV/Tuning
    try:
        # Verificar mínimo de puntos totales
        if len(df_prophet_train) < min_data_points_total:
            raise ValueError(f"Datos de entrenamiento insuficientes ({len(df_prophet_train)})")

        # Verificar rango de fechas suficiente para la configuración de CV
        min_data_points_cv_days = pd.Timedelta(CV_INITIAL).days + pd.Timedelta(CV_HORIZON).days
        data_range_days = (df_prophet_train['ds'].max() - df_prophet_train['ds'].min()).days
        if data_range_days < min_data_points_cv_days:
             print(f"ADVERTENCIA: Rango de datos ({data_range_days} días) insuficiente para CV/Tuning con horizonte {CV_HORIZON}. Se omitirá CV/Tuning.")
             can_run_cv_tuning = False

    except Exception as e_cv_check:
        print(f"ERROR verificando suficiencia de datos para {etf_symbol}: {e_cv_check}")
        results_prophet[etf_symbol]['error'] = f"Error verificando datos: {e_cv_check}"
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect()
        continue # Saltar al siguiente ETF

    last_train_date = df_prophet_train['ds'].max()
    results_prophet[etf_symbol]['last_real_date'] = last_train_date # Guardar última fecha para comparación
    print(f"Datos de entrenamiento listos. Última fecha: {last_train_date.strftime('%Y-%m-%d')}. Puntos: {len(df_prophet_train)}")

    # --- 4.2 ## NUEVO ## : Ajuste de Hiperparámetros (Tuning) Opcional ---
    best_params_for_etf = PROPHET_BASE_PARAMS.copy() # Empezar con los base
    tuning_ran_successfully = False
    best_cv_metric_value = None

    # Solo ejecutar si está activado, hay suficientes datos, y CV está habilitado globalmente
    if PERFORM_HYPERPARAMETER_TUNING and EJECUTAR_CROSS_VALIDATION and can_run_cv_tuning:
        print(f"\n--- Iniciando Ajuste de Hiperparámetros (Tuning) para {etf_symbol} ---")
        print(f"Métrica a optimizar: {CV_METRIC_TO_OPTIMIZE}")
        print(f"Grid de parámetros: {PARAM_GRID}")

        # Generar todas las combinaciones de parámetros del grid
        all_params = [dict(zip(PARAM_GRID.keys(), v)) for v in itertools.product(*PARAM_GRID.values())]
        tuning_results = [] # Lista para guardar {'params': ..., 'metric': ...}

        print(f"Probando {len(all_params)} combinaciones de parámetros...")
        for i, params_to_try in enumerate(all_params):
            current_trial_params = {**PROPHET_BASE_PARAMS, **params_to_try} # Combinar base + grid
            print(f"  Prueba {i+1}/{len(all_params)}: {params_to_try}", end="")
            try:
                # Instanciar modelo temporal para CV
                m_tune = Prophet(**current_trial_params, holidays=CUSTOM_HOLIDAYS)
                # Añadir regresores si existen
                if REGRESSOR_COLUMNS:
                    for regressor in REGRESSOR_COLUMNS:
                        m_tune.add_regressor(regressor)
                # Entrenar modelo temporal
                m_tune.fit(df_prophet_train)
                # Ejecutar Cross-Validation
                df_cv_tune = cross_validation(m_tune, initial=CV_INITIAL, period=CV_PERIOD, horizon=CV_HORIZON, parallel="processes", disable_tqdm=True)
                # Calcular métricas de rendimiento
                df_p_tune = performance_metrics(df_cv_tune)
                # Obtener el valor promedio de la métrica objetivo
                metric_value = df_p_tune[CV_METRIC_TO_OPTIMIZE].mean()
                tuning_results.append({'params': params_to_try, 'metric': metric_value})
                print(f" -> {CV_METRIC_TO_OPTIMIZE}={metric_value:.4f}")
                # Limpiar memoria del modelo temporal
                del m_tune, df_cv_tune, df_p_tune; gc.collect()

            except Exception as e_tune:
                print(f" -> ERROR: {e_tune}")
                # Guardar resultado como infinito para que no sea elegido como el mejor
                tuning_results.append({'params': params_to_try, 'metric': float('inf')})
                # Limpieza adicional en caso de error
                if 'm_tune' in locals(): del m_tune
                if 'df_cv_tune' in locals(): del df_cv_tune
                if 'df_p_tune' in locals(): del df_p_tune
                gc.collect()

        # Encontrar la mejor combinación (menor métrica)
        if tuning_results:
            best_run = min(tuning_results, key=lambda x: x['metric'])
            # Asegurarse que el mejor run no fue un error (inf)
            if best_run['metric'] != float('inf'):
                best_params_for_etf.update(best_run['params']) # Actualizar los params base con los mejores del grid
                best_cv_metric_value = best_run['metric']
                print(f"\nMejores parámetros encontrados para {etf_symbol}: {best_run['params']}")
                print(f"Mejor valor de {CV_METRIC_TO_OPTIMIZE} (promedio CV): {best_cv_metric_value:.4f}")
                tuning_ran_successfully = True
            else:
                print("\nEl Tuning falló para todas las combinaciones. Se usarán los parámetros base.")
        else:
            print("\nNo se obtuvieron resultados del Tuning. Se usarán los parámetros base.")

    else:
        # Mensajes indicando por qué no se hizo tuning
        if not PERFORM_HYPERPARAMETER_TUNING: print("\nAjuste de Hiperparámetros OMITIDO (desactivado por configuración).")
        elif not EJECUTAR_CROSS_VALIDATION: print("\nAjuste de Hiperparámetros OMITIDO (Cross-Validation global desactivada).")
        elif not can_run_cv_tuning: print("\nAjuste de Hiperparámetros OMITIDO (datos insuficientes para la configuración de CV).")

    # Guardar los parámetros finales usados (base o tuneados)
    results_prophet[etf_symbol]['params'] = best_params_for_etf.copy()
    print(f"Parámetros finales a usar para el modelo de {etf_symbol}: {best_params_for_etf}")


    # --- 4.3 Entrenar Modelo FINAL --- ## MODIFICADO ##
    print(f"\nEntrenando modelo FINAL para {etf_symbol}...")
    model_final = Prophet(**best_params_for_etf, holidays=CUSTOM_HOLIDAYS)

    # Añadir regresores si se especificaron
    if REGRESSOR_COLUMNS:
        print(f"Añadiendo regresores al modelo final: {REGRESSOR_COLUMNS}")
        for regressor in REGRESSOR_COLUMNS:
            model_final.add_regressor(regressor)

    # Añadir feriados de USA (si no se usan CUSTOM_HOLIDAYS que los incluyan)
    # Descomentar si es necesario y CUSTOM_HOLIDAYS es None
    # try:
    #     if CUSTOM_HOLIDAYS is None:
    #         model_final.add_country_holidays(country_name='US')
    #         print("Feriados de USA añadidos al modelo final.")
    # except Exception as e_hol:
    #     print(f"Advertencia: No se pudieron añadir feriados de USA: {e_hol}")

    try:
        model_final.fit(df_prophet_train)
        print("Entrenamiento del modelo final completado.")
    except Exception as e_fit_final:
        print(f"ERROR FATAL al entrenar el modelo FINAL para {etf_symbol}: {e_fit_final}")
        traceback.print_exc() # Imprimir detalles del error
        results_prophet[etf_symbol]['error'] = f"Error entrenamiento final: {e_fit_final}"
        # Limpieza robusta en caso de fallo
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        gc.collect()
        continue # Saltar al siguiente ETF

    # --- 4.4 Resumen CV/Tuning --- ## MODIFICADO ##
    # Ya no ejecutamos CV separada aquí. Guardamos un resumen del tuning.
    cv_summary_message = "CV/Tuning no ejecutado o fallido."
    if tuning_ran_successfully:
        cv_summary_message = f"Tuning OK. Mejor {CV_METRIC_TO_OPTIMIZE}: {best_cv_metric_value:.4f}"
    elif not PERFORM_HYPERPARAMETER_TUNING:
        cv_summary_message = "Tuning desactivado."
    elif not EJECUTAR_CROSS_VALIDATION:
         cv_summary_message = "Tuning omitido (CV global desactivada)."
    elif not can_run_cv_tuning:
         cv_summary_message = "Tuning omitido (datos insuficientes)."

    results_prophet[etf_symbol]['cv_metrics_summary'] = cv_summary_message # Guardar el mensaje resumen


    # --- 4.5 Generar Forecast Final --- ## MODIFICADO ##
    print(f"\nGenerando Forecast Final para {etf_symbol} hasta {FUTURE_END_DATE}...")
    forecast_final_scale = None # Variable para el forecast en escala original
    future_df_final = None # Variable para el dataframe futuro

    try:
        # Crear dataframe futuro base con las fechas
        target_end_date = pd.to_datetime(FUTURE_END_DATE)
        freq = 'B' # Frecuencia de días hábiles (Business days)
        # Asegurar que empezamos un día después del último dato de entrenamiento
        future_dates = pd.date_range(start=last_train_date + pd.Timedelta(days=1), end=target_end_date, freq=freq)

        if future_dates.empty:
             raise ValueError(f"No se generaron fechas futuras (quizás {FUTURE_END_DATE} es anterior a {last_train_date.strftime('%Y-%m-%d')})")

        future_df_final = pd.DataFrame({'ds': future_dates})
        future_df_final = future_df_final[future_df_final['ds'] <= target_end_date] # Asegurar no pasar la fecha final

        # --- ## NUEVO ## : Añadir valores futuros para los regresores ---
        if REGRESSOR_COLUMNS:
            print(f"Preparando valores futuros para regresores: {REGRESSOR_COLUMNS}")
            try:
                # Intentar cargar datos completos que podrían tener los regresores futuros
                df_actual_full = pd.read_csv(ACTUAL_DATA_FILE)
                df_actual_full['ds'] = pd.to_datetime(df_actual_full['price_date']) # Usar 'ds' temporalmente
                df_reg_future_values = df_actual_full.loc[df_actual_full['fund_symbol'] == etf_symbol, ['ds'] + REGRESSOR_COLUMNS].copy()

                # Hacer merge con el dataframe futuro
                future_df_final = pd.merge(future_df_final, df_reg_future_values, on='ds', how='left')

                # Rellenar posibles NaNs en los regresores futuros
                for col in REGRESSOR_COLUMNS:
                    # Primero, ffill (usar el último valor conocido)
                    future_df_final[col] = future_df_final[col].ffill()
                    # Si aún quedan NaNs (al principio), usar el último valor del set de entrenamiento
                    if future_df_final[col].isnull().any():
                        last_known_regressor_value = df_prophet_train[col].iloc[-1]
                        future_df_final[col].fillna(last_known_regressor_value, inplace=True)
                        print(f"  ADVERTENCIA: NaNs en regresor futuro '{col}' rellenados con último valor conocido ({last_known_regressor_value:.4f}).")

                    # Verificar si aún quedan NaNs (no debería pasar, pero por seguridad)
                    if future_df_final[col].isnull().any():
                         raise ValueError(f"No se pudieron obtener/rellenar valores futuros para el regresor '{col}'.")

                del df_actual_full, df_reg_future_values # Limpiar memoria
                print("Valores futuros de regresores añadidos.")

            except FileNotFoundError:
                 raise ValueError(f"No se encontró el archivo '{ACTUAL_DATA_FILE}' necesario para obtener regresores futuros.")
            except Exception as e_reg_fut:
                 raise ValueError(f"Error al preparar regresores futuros: {e_reg_fut}")

        # Realizar la predicción con el modelo final
        print(f"Realizando predicción con {len(future_df_final)} fechas futuras...")
        if future_df_final.empty:
             raise ValueError("El DataFrame futuro está vacío antes de la predicción.")

        forecast_raw = model_final.predict(future_df_final)
        print("Predicción completada.")

        # Copiar el forecast para trabajar sobre él
        forecast_final_scale = forecast_raw.copy()

        # --- ## NUEVO ## : Revertir Transformación Logarítmica si se aplicó ---
        if is_log_transformed_etf:
            print(f"Revirtiendo transformación logarítmica para {etf_symbol}...")
            cols_to_exp = ['yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']
            # Incluir componentes de estacionalidad si existen en el forecast_raw
            for col in forecast_raw.columns:
                 if 'seasonality' in col or col in ('weekly', 'yearly', 'daily', 'multiplicative_terms', 'additive_terms'): # Añadir otros componentes si es necesario
                     if col not in cols_to_exp:
                         cols_to_exp.append(col)
                 # Incluir regresores si existen
                 if col in REGRESSOR_COLUMNS and col + '_lower' not in cols_to_exp: # Revertir efecto del regresor
                    cols_to_exp.extend([col, col + '_lower', col + '_upper'])


            cols_reverted = []
            for col in cols_to_exp:
                if col in forecast_final_scale.columns:
                    try:
                        # Evitar revertir valores extremadamente bajos que puedan dar inf o NaN
                        # forecast_final_scale[col] = np.exp(np.clip(forecast_final_scale[col], a_min=-700, a_max=700)) # Clip para evitar overflow
                        forecast_final_scale[col] = np.exp(forecast_final_scale[col])
                        cols_reverted.append(col)
                    except Exception as e_exp:
                        print(f"  ERROR al revertir (np.exp) columna '{col}': {e_exp}. Dejando como está.")
                        # Podrías optar por poner NaN si falla la reversión: forecast_final_scale[col] = np.nan
            print(f"Columnas revertidas a escala original: {cols_reverted}")

        # Calcular ancho del intervalo de predicción en escala final
        if 'yhat_upper' in forecast_final_scale.columns and 'yhat_lower' in forecast_final_scale.columns:
             forecast_final_scale['prediction_interval_width'] = forecast_final_scale['yhat_upper'] - forecast_final_scale['yhat_lower']

        # Guardar el forecast relevante (en escala original)
        cols_to_save = ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width']
        cols_present_in_forecast = [c for c in cols_to_save if c in forecast_final_scale.columns]
        results_prophet[etf_symbol]['forecast'] = forecast_final_scale[cols_present_in_forecast].copy()

        print(f"Forecast final (escala original) generado y guardado. Última fecha: {forecast_final_scale['ds'].max().strftime('%Y-%m-%d')}.")

    except Exception as e_pred:
        print(f"ERROR FATAL generando el forecast final para {etf_symbol}: {e_pred}")
        traceback.print_exc() # Imprimir detalles
        results_prophet[etf_symbol]['error_forecast'] = str(e_pred)
        # Limpieza robusta en caso de fallo en predicción
        if 'model_final' in locals(): del model_final
        if 'df_etf' in locals(): del df_etf
        if 'df_prophet_train' in locals(): del df_prophet_train
        if 'future_df_final' in locals(): del future_df_final
        if 'forecast_raw' in locals(): del forecast_raw
        if 'forecast_final_scale' in locals(): del forecast_final_scale
        gc.collect()
        continue # Saltar al siguiente ETF


    # --- 4.6 Visualización Inicial (Opcional - Comentada por defecto) ---
    # print(f"\nGenerando gráficos iniciales para {etf_symbol}...")
    # try:
    #     if forecast_final_scale is not None: # Usar el forecast en escala original
    #         # Nota: model_final.plot() usará la escala en la que fue entrenado (log si aplica).
    #         # Para plotear en escala original, puede ser mejor hacerlo manualmente o ajustar ejes.
    #         fig1 = model_final.plot(forecast_raw) # Plotea sobre el forecast antes de revertir log
    #         plt.axvline(last_train_date, color='gray', linestyle=':', lw=2, label='Último Dato Entrenamiento')
    #         y_label = "Valor (y)" if not is_log_transformed_etf else "Valor (log(y))"
    #         plt.title(f'Predicción Prophet Base para {etf_symbol}')
    #         plt.xlabel("Fecha"); plt.ylabel(y_label)
    #         plt.legend(); plt.tight_layout(); plt.show()

    #         fig2 = model_final.plot_components(forecast_raw) # Componentes también en escala de entrenamiento
    #         plt.suptitle(f'Componentes del Modelo Prophet para {etf_symbol}', y=1.02)
    #         plt.tight_layout(); plt.show()
    #         print(f"Gráficos iniciales generados.")
    #     else:
    #         print("No se generó forecast, omitiendo gráficos iniciales.")
    # except Exception as e_plot_init:
    #     print(f"Error generando gráficos iniciales: {e_plot_init}")

    # --- 4.7 Limpieza de Memoria --- ## MODIFICADO ## (Más completo)
    print(f"Limpiando memoria para {etf_symbol}...")
    del model_final, df_etf, df_prophet_train
    if 'future_df_final' in locals(): del future_df_final
    if 'forecast_raw' in locals(): del forecast_raw
    if 'forecast_final_scale' in locals(): del forecast_final_scale
    # Variables de tuning
    if 'tuning_results' in locals(): del tuning_results
    if 'best_run' in locals(): del best_run
    if 'all_params' in locals(): del all_params
    # Variables de CV si se usaron en tuning
    # No es necesario borrar m_tune, df_cv_tune, df_p_tune aquí porque se borran dentro del bucle de tuning
    gc.collect()
    print("Limpieza completada.")

    print(f"\n{'='*15} Procesamiento de {etf_symbol} completado {'='*15}")
    print("-" * 70)


print("\n###########################################################")
print("   ANÁLISIS COMPLETADO PARA TODOS LOS ETFs SELECCIONADOS")
print("###########################################################")

# --- 5. MOSTRAR RESUMEN FINAL DE RESULTADOS DEL ANÁLISIS --- ## MODIFICADO ## (para mostrar resumen CV/Tuning)
print("\n--- Resumen de Resultados del Análisis ---")
for etf, data in results_prophet.items():
    print(f"\n--- ETF: {etf} ---")
    print(f"  Transformación Log Aplicada: {data.get('log_transform_applied', 'N/D')}")
    print(f"  Parámetros Finales Usados: {data.get('params', 'N/A')}")
    if 'error' in data:
        print(f"  ERROR PRINCIPAL: {data['error']}")
        continue # No mostrar más si hubo error temprano

    # Mostrar resumen de CV/Tuning
    print("\n  Resultado Cross-Validation / Tuning:")
    cv_summary = data.get('cv_metrics_summary', 'No disponible')
    print(f"    {cv_summary}") # Mostrar el string resumen

    # Mostrar ejemplo de predicciones finales
    print("\n  Ejemplo Predicciones Finales (Últimas 5 - Escala Original):")
    if 'forecast' in data and not data['forecast'].empty:
        # Asegurarse que las columnas existen antes de intentar mostrarlas
        cols_to_display = [c for c in ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'prediction_interval_width'] if c in data['forecast'].columns]
        if cols_to_display:
             final_preds_display = data['forecast'][cols_to_display].tail()
             print(final_preds_display.round(4).to_string(index=False))
        else:
             print("    No se encontraron columnas estándar de forecast para mostrar.")
    elif 'error_forecast' in data:
         print(f"    ERROR en Forecast: {data['error_forecast']}")
    else:
         print("    No se encontraron datos de forecast.")
    print("-" * 40)
print("\n--- Fin del Resumen del Análisis ---")


# --- 6. COMPARACIÓN CON DATOS REALES ACTUALIZADOS ---
# (Carga y preparación inicial sin cambios respecto a Script 1)

print("\n\n###########################################################")
print(f"      COMPARACIÓN CON DATOS REALES DESDE '{ACTUAL_DATA_FILE}'")
print("###########################################################")

df_actual_new = None # Inicializar
comparison_possible = False
try:
    # Cargar los datos reales completos/actualizados
    df_actual_new = pd.read_csv(ACTUAL_DATA_FILE)
    print(f"Datos reales cargados desde '{ACTUAL_DATA_FILE}'.")

    # --- Preparación de Datos Reales ---
    DATE_COL_ACTUAL = 'price_date'
    VALUE_COL_ACTUAL = 'adj_close'
    SYMBOL_COL_ACTUAL = 'fund_symbol'
    required_actual_cols = [DATE_COL_ACTUAL, VALUE_COL_ACTUAL, SYMBOL_COL_ACTUAL]
    if not all(col in df_actual_new.columns for col in required_actual_cols):
        raise ValueError(f"'{ACTUAL_DATA_FILE}' debe contener: {required_actual_cols}")

    df_actual_new[DATE_COL_ACTUAL] = pd.to_datetime(df_actual_new[DATE_COL_ACTUAL])
    df_actual_new = df_actual_new.rename(columns={
        DATE_COL_ACTUAL: 'ds',
        VALUE_COL_ACTUAL: 'y_actual',
        SYMBOL_COL_ACTUAL: 'fund_symbol'
    })
    # Eliminar NaNs en columnas clave para la comparación
    df_actual_new.dropna(subset=['ds', 'y_actual', 'fund_symbol'], inplace=True)
    df_actual_new = df_actual_new.sort_values(by=['fund_symbol', 'ds'])
    print("Datos reales preparados (columnas renombradas, NaNs eliminados).")
    comparison_possible = True

except FileNotFoundError:
    print(f"ERROR: El archivo '{ACTUAL_DATA_FILE}' no se encontró.")
    print("Asegúrate de que el archivo exista y la ruta sea correcta.")
    comparison_possible = False
except Exception as e:
    print(f"Error al cargar o preparar '{ACTUAL_DATA_FILE}': {e}")
    traceback.print_exc()
    comparison_possible = False

# --- Generar Gráficos de Comparación (MANUALMENTE con EJE Y LOG) --- ## MODIFICADO ##
if comparison_possible and df_actual_new is not None:
    print("\n--- Generando Gráficos de Comparación (Predicción vs Realidad Futura - EJE Y LOG) ---") # Título actualizado

    for etf_symbol, results in results_prophet.items():
        print(f"\n--- Comparando para ETF: {etf_symbol} ---")

        # Verificar que tenemos forecast y la última fecha de entrenamiento para este ETF
        if 'forecast' in results and not results['forecast'].empty and 'last_real_date' in results:
            forecast_data = results['forecast'].copy()
            last_train_date_plot = results['last_real_date'] # Renombrado para claridad en plot

            # Filtrar datos reales actualizados para este ETF y para fechas POSTERIORES al entrenamiento
            actual_data_etf_plot = df_actual_new[
                (df_actual_new['fund_symbol'] == etf_symbol) &
                (df_actual_new['ds'] > last_train_date_plot)
            ].copy()

            if actual_data_etf_plot.empty:
                print(f"No se encontraron datos reales en '{ACTUAL_DATA_FILE}' para {etf_symbol} después de {last_train_date_plot.strftime('%Y-%m-%d')}.")
                continue

            print(f"Encontrados {len(actual_data_etf_plot)} puntos reales futuros para comparar.")

            # --- Creación del Gráfico Manual ---
            try:
                print(f"Preparando gráfico manual para {etf_symbol}")
                # Recuperar datos históricos 'y' del df de análisis original para plotearlos
                # Renombrar a 'y_hist_orig' para evitar colisión con 'y' (log) si se usó
                historical_data_plot = df_analysis.loc[
                    df_analysis['fund_symbol'] == etf_symbol, ['price_date', 'adj_close']
                ].rename(columns={'price_date': 'ds', 'adj_close': 'y_hist_orig'}).dropna()

                # Asegurarse que las columnas necesarias están en el forecast
                required_forecast_cols = ['ds', 'yhat', 'yhat_lower', 'yhat_upper']
                if not all(col in forecast_data.columns for col in required_forecast_cols):
                     print(f"ADVERTENCIA: Faltan columnas en el forecast para {etf_symbol}. Columnas presentes: {forecast_data.columns.tolist()}")
                     # Crear columnas dummy si faltan para evitar error en merge, aunque el plot será incompleto
                     for col in required_forecast_cols:
                         if col not in forecast_data.columns: forecast_data[col] = np.nan


                # Unir: forecast (predicciones futuras) + datos reales futuros + datos históricos
                # 1. Merge forecast con datos reales futuros
                comparison_df_plot = pd.merge(forecast_data[required_forecast_cols],
                                              actual_data_etf_plot[['ds', 'y_actual']],
                                              on='ds', how='left')
                # 2. Merge resultado con datos históricos
                # Usar outer merge para incluir todo el rango de fechas (histórico + futuro)
                comparison_df_plot_full = pd.merge(historical_data_plot[['ds', 'y_hist_orig']],
                                                   comparison_df_plot,
                                                   on='ds', how='outer')

                # Ordenar por fecha para el gráfico
                comparison_df_plot_full = comparison_df_plot_full.sort_values(by='ds').reset_index(drop=True)
                print(f"DataFrame de comparación preparado con {len(comparison_df_plot_full)} filas.")

                # Crear figura y ejes
                fig_comp, ax_comp = plt.subplots(figsize=(14, 7)) # Tamaño un poco más grande

                # --- Graficar elementos ---

                # a) Datos históricos observados (puntos negros) - Filtrar > 0 para log
                hist_plot_data = comparison_df_plot_full.dropna(subset=['y_hist_orig'])
                hist_plot_data = hist_plot_data[hist_plot_data['y_hist_orig'] > 0]
                if not hist_plot_data.empty:
                    ax_comp.plot(hist_plot_data['ds'].dt.to_pydatetime(), hist_plot_data['y_hist_orig'],
                                 'k.', markersize=3, label='Datos Observados (Históricos)')

                # b) Predicción (línea azul) - Filtrar > 0 para log
                forecast_plot_data = comparison_df_plot_full.dropna(subset=['yhat'])
                forecast_plot_data = forecast_plot_data[forecast_plot_data['yhat'] > 0]
                if not forecast_plot_data.empty:
                    ax_comp.plot(forecast_plot_data['ds'].dt.to_pydatetime(), forecast_plot_data['yhat'],
                                 ls='-', c='#0072B2', label='Predicción (yhat)')

                # c) Intervalo de incertidumbre (área sombreada azul) - Filtrar > 0 para log
                uncertainty_plot_data = comparison_df_plot_full.dropna(subset=['yhat_lower', 'yhat_upper'])
                uncertainty_plot_data = uncertainty_plot_data[(uncertainty_plot_data['yhat_lower'] > 0) & (uncertainty_plot_data['yhat_upper'] > 0)]
                if not uncertainty_plot_data.empty:
                     ax_comp.fill_between(uncertainty_plot_data['ds'].dt.to_pydatetime(),
                                          uncertainty_plot_data['yhat_lower'], uncertainty_plot_data['yhat_upper'],
                                          color='#0072B2', alpha=0.2, label='Intervalo Incertidumbre')

                # d) Datos reales futuros (puntos rojos) - Filtrar > 0 para log
                actual_plot_data = comparison_df_plot_full.dropna(subset=['y_actual'])
                actual_plot_data = actual_plot_data[actual_plot_data['y_actual'] > 0]
                if not actual_plot_data.empty:
                    ax_comp.plot(actual_plot_data['ds'].dt.to_pydatetime(), actual_plot_data['y_actual'],
                                 'r.', markersize=5, label='Datos Reales (Futuros)')
                else:
                     print(f"Advertencia: No hay datos 'y_actual' positivos para graficar en escala log.")

                # e) Línea vertical indicando fin del entrenamiento
                ax_comp.axvline(last_train_date_plot, color='grey', linestyle=':', lw=2, label='Fin Entrenamiento')

                # --- Configuración del Gráfico ---
                ax_comp.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
                ax_comp.grid(True, which='minor', c='gray', ls=':', lw=0.5, alpha=0.1) # Añadir grid menor
                ax_comp.set_xlabel('Fecha')
                # ## MODIFICADO ## : Etiqueta y escala del eje Y
                ax_comp.set_ylabel('Valor (adj_close) - Escala Logarítmica')
                ax_comp.set_title(f'Comparación Predicción vs Realidad para {etf_symbol} (Escala Log)')
                ax_comp.set_yscale('log') # <-- Establecer escala logarítmica
                print("Eje Y configurado a escala logarítmica.")

                # Ajustar límites del eje X para enfocar en el período de interés
                # Mostrar un poco antes del fin del entrenamiento y hasta el final del forecast/realidad
                days_before = 60 # Mostrar ~2 meses antes
                plot_start_date = last_train_date_plot - pd.Timedelta(days=days_before)
                plot_end_date = comparison_df_plot_full['ds'].max()
                # Añadir un pequeño margen al final
                if pd.notna(plot_end_date):
                    ax_comp.set_xlim([plot_start_date, plot_end_date + pd.Timedelta(days=10)])
                else: # Si no hay fechas futuras, usar la fecha de entrenamiento como final
                     ax_comp.set_xlim([plot_start_date, last_train_date_plot + pd.Timedelta(days=10)])


                # Añadir leyenda
                ax_comp.legend()

                # Mostrar
                plt.tight_layout()
                plt.show()
                print(f"Gráfico de comparación (log) generado manualmente para {etf_symbol}.")

            except Exception as e_plot:
                print(f"Error generando el gráfico de comparación manual para {etf_symbol}: {e_plot}")
                traceback.print_exc() # Mostrar error detallado

        elif 'last_real_date' not in results:
             print(f"Saltando gráfico para {etf_symbol}: Falta 'last_real_date' en los resultados.")
        else:
             print(f"Saltando gráfico para {etf_symbol}: No hay 'forecast' disponible en los resultados.")

        # Limpieza robusta después de cada gráfico ## MODIFICADO ##
        if 'forecast_data' in locals(): del forecast_data
        if 'actual_data_etf_plot' in locals(): del actual_data_etf_plot
        if 'historical_data_plot' in locals(): del historical_data_plot
        if 'comparison_df_plot' in locals(): del comparison_df_plot
        if 'comparison_df_plot_full' in locals(): del comparison_df_plot_full
        if 'hist_plot_data' in locals(): del hist_plot_data
        if 'forecast_plot_data' in locals(): del forecast_plot_data
        if 'uncertainty_plot_data' in locals(): del uncertainty_plot_data
        if 'actual_plot_data' in locals(): del actual_plot_data
        if 'fig_comp' in locals(): plt.close(fig_comp) # Cerrar figura explícitamente
        gc.collect()

    print("\n--- Fin de la Comparación ---")
else:
    print("\nNo se generaron gráficos de comparación debido a problemas al cargar o procesar los datos reales.")

print("\n--- Script Finalizado ---")
```
